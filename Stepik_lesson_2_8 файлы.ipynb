```python
# –®–∞–≥ 2_8_1

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –†–∞–±–æ—Ç–∞ —Å —Ñ–∞–π–ª–∞–º–∏ –≤ Python

# Python –ø–æ–∑–≤–æ–ª—è–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –∫–∞–∫ —Å —Ç–µ–∫—Å—Ç–æ–≤—ã–º–∏ (txt, csv, json, xml –∏ —Ç.–ø.), —Ç–∞–∫ –∏ —Å –±–∏–Ω–∞—Ä–Ω—ã–º–∏ (–∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è, –º—É–∑—ã–∫–∞ –∏ —Ç.–ø.) —Ñ–∞–π–ª–∞–º–∏.

# --- –û—Ç–∫—Ä—ã—Ç–∏–µ —Ñ–∞–π–ª–∞ ---

# –ü—Ä–µ–∂–¥–µ —á–µ–º –Ω–∞—á–∞—Ç—å —Ä–∞–±–æ—Ç—É —Å —Ñ–∞–π–ª–æ–º ‚Äî –µ–≥–æ –Ω—É–∂–Ω–æ –æ—Ç–∫—Ä—ã—Ç—å:

f = open('text.txt', 'r')  # –æ—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª –¥–ª—è —á—Ç–µ–Ω–∏—è

# ‚Ä¢ –ü–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä ‚Äî –∏–º—è —Ñ–∞–π–ª–∞.
# ‚Ä¢ –í—Ç–æ—Ä–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä ‚Äî —Ä–µ–∂–∏–º –æ—Ç–∫—Ä—ã—Ç–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 'r' ‚Äî —á—Ç–µ–Ω–∏–µ).

# üìã –¢–∞–±–ª–∏—Ü–∞ —Ä–µ–∂–∏–º–æ–≤:

| –†–µ–∂–∏–º | –û–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ                                       |
|-------|---------------------------------------------------|
| 'r'   | –æ—Ç–∫—Ä—ã—Ç–∏–µ –Ω–∞ —á—Ç–µ–Ω–∏–µ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)                 |
| 'w'   | –æ—Ç–∫—Ä—ã—Ç–∏–µ –Ω–∞ –∑–∞–ø–∏—Å—å (—Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —É–¥–∞–ª—è–µ—Ç—Å—è –∏–ª–∏ —Å–æ–∑–¥–∞—ë—Ç—Å—è –Ω–æ–≤—ã–π —Ñ–∞–π–ª) |
| 'x'   | –æ—Ç–∫—Ä—ã—Ç–∏–µ –Ω–∞ –∑–∞–ø–∏—Å—å, —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ —Ñ–∞–π–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç (–∏–Ω–∞—á–µ ‚Äî –æ—à–∏–±–∫–∞) |
| 'a'   | –æ—Ç–∫—Ä—ã—Ç–∏–µ –Ω–∞ –¥–æ–∑–∞–ø–∏—Å—å (–∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –≤ –∫–æ–Ω–µ—Ü —Ñ–∞–π–ª–∞) |
| 'b'   | –æ—Ç–∫—Ä—ã—Ç–∏–µ –≤ –¥–≤–æ–∏—á–Ω–æ–º —Ä–µ–∂–∏–º–µ                        |
| 't'   | –æ—Ç–∫—Ä—ã—Ç–∏–µ –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º —Ä–µ–∂–∏–º–µ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)         |
| '+'   | –æ—Ç–∫—Ä—ã—Ç–∏–µ –Ω–∞ —á—Ç–µ–Ω–∏–µ –∏ –∑–∞–ø–∏—Å—å                       |

# ‚ö†Ô∏è –í–∞–∂–Ω–æ:
# –ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è `f` ‚Äî —ç—Ç–æ **–Ω–µ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ñ–∞–π–ª–∞**, –∞ **—Å—Å—ã–ª–∫–∞ –Ω–∞ —Ñ–∞–π–ª** (—Ñ–∞–π–ª-—Ö—ç–Ω–¥–ª–µ—Ä, –¥–µ—Å–∫—Ä–∏–ø—Ç–æ—Ä).
# –ß—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ ‚Äî –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥—ã —Ç–∏–ø–∞ `.read()`, `.readline()`, `.readlines()` –∏–ª–∏ —Ü–∏–∫–ª `for line in f`.

# --- –ö–æ–¥–∏—Ä–æ–≤–∫–∞ ---

# –ü—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å —Ç–µ–∫—Å—Ç–æ–º –Ω–∞ —Ä—É—Å—Å–∫–æ–º –∏–ª–∏ –¥—Ä—É–≥–∏—Ö —è–∑—ã–∫–∞—Ö ‚Äî —É–∫–∞–∑—ã–≤–∞–π—Ç–µ –∫–æ–¥–∏—Ä–æ–≤–∫—É:

f = open('text.txt', 'r', encoding='utf-8')  # –¥–ª—è –∫–∏—Ä–∏–ª–ª–∏—Ü—ã, —ç–º–æ–¥–∑–∏ –∏ —Ç.–ø.

# –ë–µ–∑ —è–≤–Ω–æ–≥–æ —É–∫–∞–∑–∞–Ω–∏—è ‚Äî Python –º–æ–∂–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–∏—Å—Ç–µ–º–Ω—É—é –∫–æ–¥–∏—Ä–æ–≤–∫—É (–Ω–∞–ø—Ä–∏–º–µ—Ä, cp1251 –Ω–∞ Windows), —á—Ç–æ –ø—Ä–∏–≤–µ–¥—ë—Ç –∫ –æ—à–∏–±–∫–µ.

# --- –ó–∞–∫—Ä—ã—Ç–∏–µ —Ñ–∞–π–ª–∞ ---

# –ü–æ—Å–ª–µ –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è —Ä–∞–±–æ—Ç—ã —Å —Ñ–∞–π–ª–æ–º –µ–≥–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∑–∞–∫—Ä—ã—Ç—å:

f.close()

# –ü–æ—á–µ–º—É?
#   ‚Ä¢ –°–æ—Ö—Ä–∞–Ω—è–µ—Ç –∏–∑–º–µ–Ω–µ–Ω–∏—è (–µ—Å–ª–∏ –±—ã–ª–∏).
#   ‚Ä¢ –û—Å–≤–æ–±–æ–∂–¥–∞–µ—Ç —Ä–µ—Å—É—Ä—Å—ã.
#   ‚Ä¢ –ü—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–∞–µ—Ç –±–ª–æ–∫–∏—Ä–æ–≤–∫—É —Ñ–∞–π–ª–∞ –¥—Ä—É–≥–∏–º–∏ –ø—Ä–æ–≥—Ä–∞–º–º–∞–º–∏.
#   ‚Ä¢ –ò–∑–±–µ–≥–∞–µ—Ç —É—Ç–µ—á–µ–∫ –ø–∞–º—è—Ç–∏.

# üí° –°–æ–≤–µ—Ç: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `with` ‚Äî –æ–Ω –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –∑–∞–∫—Ä–æ–µ—Ç —Ñ–∞–π–ª –¥–∞–∂–µ –ø—Ä–∏ –æ—à–∏–±–∫–µ:

with open('text.txt', 'r', encoding='utf-8') as f:
    content = f.read()
# —Ñ–∞–π–ª –∑–∞–∫—Ä—ã—Ç –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏!
```

```python
# –®–∞–≥ 2_8_2

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –ß—Ç–µ–Ω–∏–µ –∏–∑ —Ñ–∞–π–ª–∞

# –ü–æ—Å–ª–µ –æ—Ç–∫—Ä—ã—Ç–∏—è —Ñ–∞–π–ª–∞ –Ω–∞ —á—Ç–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, `open('file.txt', 'r')`) –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç—Ä–∏ –æ—Å–Ω–æ–≤–Ω—ã—Ö –º–µ—Ç–æ–¥–∞ –¥–ª—è —á—Ç–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–æ–≤–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏:

# 1. `read()` ‚Äî —á–∏—Ç–∞–µ—Ç **–≤–µ—Å—å —Ñ–∞–π–ª —Ü–µ–ª–∏–∫–æ–º** –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É.
# 2. `readline()` ‚Äî —á–∏—Ç–∞–µ—Ç **–æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É** (–≤–∫–ª—é—á–∞—è —Å–∏–º–≤–æ–ª –ø–µ—Ä–µ–Ω–æ—Å–∞ `\n`).
# 3. `readlines()` ‚Äî —á–∏—Ç–∞–µ—Ç **–≤–µ—Å—å —Ñ–∞–π–ª** –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ (–∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è `\n`).

# ‚ö†Ô∏è –í–∞–∂–Ω–æ:
# –í—Å–µ —ç—Ç–∏ –º–µ—Ç–æ–¥—ã –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç —Å—Ç—Ä–æ–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ **–≤–∫–ª—é—á–∞—é—Ç —Å–∏–º–≤–æ–ª –ø–µ—Ä–µ–Ω–æ—Å–∞ —Å—Ç—Ä–æ–∫–∏ `\n`** (–∫—Ä–æ–º–µ –ø–æ—Å–ª–µ–¥–Ω–µ–π —Å—Ç—Ä–æ–∫–∏, –µ—Å–ª–∏ –æ–Ω–∞ –Ω–µ –æ–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –ø–µ—Ä–µ–≤–æ–¥–æ–º —Å—Ç—Ä–æ–∫–∏).

# --- –ü—Ä–∏–º–µ—Ä—ã ---

# –ß–∏—Ç–∞–µ–º –≤–µ—Å—å —Ñ–∞–π–ª —Ü–µ–ª–∏–∫–æ–º:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
text = f.read()
print(text)  # ‚Üí –≤—ã–≤–æ–¥–∏—Ç –≤—Å—ë —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ñ–∞–π–ª–∞
f.close()

# –ß–∏—Ç–∞–µ–º –ø–æ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
first_line = f.readline()  # ‚Üí 'IE_NAME;IE_ID\n'
second_line = f.readline() # ‚Üí '–û–û–û ¬´–°–†-–ì—Ä—É–ø–ø¬ª;718436\n'
f.close()

# –ß–∏—Ç–∞–µ–º –≤—Å–µ —Å—Ç—Ä–æ–∫–∏ –≤ —Å–ø–∏—Å–æ–∫:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
lines = f.readlines()  # ‚Üí —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫, –∫–∞–∂–¥–∞—è —Å \n
f.close()

# --- –ü–æ—Å—Ç—Ä–æ—á–Ω–æ–µ —á—Ç–µ–Ω–∏–µ –≤ —Ü–∏–∫–ª–µ for ---

# Python –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å —Å—Ç—Ä–æ–∫–∏ —Ñ–∞–π–ª–∞ –Ω–∞–ø—Ä—è–º—É—é, –±–µ–∑ —Å–æ–∑–¥–∞–Ω–∏—è —Å–ø–∏—Å–∫–∞:

f = open('export_file_48603.csv', 'r', encoding='utf-8')
for line in f:
    print(line, end='')  # end='' —á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –¥–≤–æ–π–Ω–æ–≥–æ –ø–µ—Ä–µ–≤–æ–¥–∞ —Å—Ç—Ä–æ–∫–∏
f.close()

# --- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–µ –º–µ—Ç–æ–¥—ã —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ ---

# –ï—Å–ª–∏ –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ **–±–µ–∑ —Å–∏–º–≤–æ–ª–æ–≤ `\n`**, –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å:

# –°–ø–æ—Å–æ–± 1: split()
f = open('export_file_48603.csv', 'r', encoding='utf-8')
content = f.read()
strings = content.split('\n')  # ‚Üí —Ä–∞–∑–¥–µ–ª—è–µ–º –ø–æ —Å–∏–º–≤–æ–ª—É –ø–µ—Ä–µ–Ω–æ—Å–∞
f.close()

# –°–ø–æ—Å–æ–± 2 (—ç–ª–µ–≥–∞–Ω—Ç–Ω–µ–µ): splitlines()
f = open('export_file_48603.csv', 'r', encoding='utf-8')
strings = f.read().splitlines()  # ‚Üí —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ –±–µ–∑ \n
f.close()

# ‚ö†Ô∏è –í–Ω–∏–º–∞–Ω–∏–µ: –µ—Å–ª–∏ –≤ —Ñ–∞–π–ª–µ –µ—Å—Ç—å –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏ –∏–ª–∏ –¥–≤–æ–π–Ω—ã–µ –ø–µ—Ä–µ–Ω–æ—Å—ã, `splitlines()` –º–æ–∂–µ—Ç –≤–µ—Ä–Ω—É—Ç—å –ø—É—Å—Ç—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã ‚Äî –∏—Ö –∏–Ω–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ.

# üí° –°–æ–≤–µ—Ç: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `with` –¥–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–∞–∫—Ä—ã—Ç–∏—è —Ñ–∞–π–ª–∞!
```

```python
# –®–∞–≥ 2_8_4

# –ó–∞–¥–∞–Ω–∏–µ: –°—É–º–º–∞ —á–∏—Å–µ–ª –∏–∑ —Ñ–∞–π–ª–∞

# –°—á–∏—Ç–∞–π—Ç–µ –∏–º—è —Ñ–∞–π–ª–∞ —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã.
# –û—Ç–∫—Ä–æ–π—Ç–µ —Ñ–∞–π–ª –∏ –ø—Ä–æ—á–∏—Ç–∞–π—Ç–µ –µ–≥–æ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ.
# –§–∞–π–ª —Å–æ–¥–µ—Ä–∂–∏—Ç –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –∑–∞–ø–∏—Å–∞–Ω–æ —Å –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏.
# –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ –ø–µ—á–∞—Ç—å —Å—É–º–º—É –≤—Å–µ—Ö —ç—Ç–∏—Ö —á–∏—Å–µ–ª.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:
#   ‚Ä¢ –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–¥–∏—Ä–æ–≤–∫—É 'utf-8'.
#   ‚Ä¢ –ù–µ –∑–∞–±—É–¥—å—Ç–µ –∑–∞–∫—Ä—ã—Ç—å —Ñ–∞–π–ª –∏–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `with`.
#   ‚Ä¢ –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞–π—Ç–µ —Å—Ç—Ä–æ–∫–∏ –≤ —á–∏—Å–ª–∞ —Å –ø–æ–º–æ—â—å—é `int()`.

# –ü—Ä–∏–º–µ—Ä—ã:

# –í–≤–æ–¥: 1.txt ‚Üí —Å–æ–¥–µ—Ä–∂–∏—Ç: 1\n2 ‚Üí –≤—ã–≤–æ–¥: 3
# –í–≤–æ–¥: 2.txt ‚Üí —Å–æ–¥–µ—Ä–∂–∏—Ç: 1\n2\n3\n4 ‚Üí –≤—ã–≤–æ–¥: 10
# –í–≤–æ–¥: 3.txt ‚Üí —Å–æ–¥–µ—Ä–∂–∏—Ç: 1\n2\n...n (—Å—É–º–º–∞ –ø–µ—Ä–≤—ã—Ö 9 —á–∏—Å–µ–ª) ‚Üí –≤—ã–≤–æ–¥: 45

a = str(input())
f = open(a, 'r', encoding = 'utf-8')
# c = f.read().split('\n')
c = f.read().splitlines()
sum = 0
for k in c:
    sum += int(k)
print(sum)
f.close()
```

```python
# –®–∞–≥ 2_8_6

"""
–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ —Å–µ–±—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–º –¥–µ–∫–∞–Ω–∞—Ç–∞. –ü—Ä–µ–ø–æ–¥–∞–≤–∞—Ç–µ–ª–∏ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è—é—Ç –≤–∞–º –ø–æ 2 —Ñ–∞–π–ª–∞, –∏–º–µ–Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö:
    sheet - —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω—É—é –≤–µ–¥–æ–º–æ—Å—Ç—å —Å —Ñ–∞–º–∏–ª–∏—è–º–∏, –∏–º–µ–Ω–∞–º–∏ –∏ –æ—Ç—á–µ—Å—Ç–≤–∞–º–∏ (–µ—Å–ª–∏ –µ—Å—Ç—å), –æ—Ü–µ–Ω–∫–∞–º–∏ –∏ —Å—Ç–∞—Ç—É—Å–∞–º–∏
    mean - —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É –ø–æ –≥—Ä—É–ø–ø–µ

–í–µ–¥–æ–º–æ—Å—Ç—å –º–æ–∂–µ—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

–ê—Ç—Ç–∏–ª–∞ 2 (—ç–∫–∑–∞–º–µ–Ω)
–ë–æ–Ω–∞–ø–∞—Ä—Ç –ù–∞–ø–æ–ª–µ–æ–Ω (–Ω–µ—è–≤–∫–∞)
–ì–∞—Å—Å–∞–Ω –ê–±–¥—É—Ä—Ä–∞—Ö–º–∞–Ω –∏–±–Ω –•–æ—Ç—Ç–∞–± 5 (–∞–≤—Ç–æ–º–∞—Ç)
–ó–∞–¥–æ–π–Ω—ã–π –ê–ª–µ–∫—Å–µ–π –í–ª–∞–¥–∏–º–∏—Ä–æ–≤–∏—á 5 (—ç–∫–∑–∞–º–µ–Ω)
–ö–æ–ª–æ–Ω–Ω–∞-–í–∞–ª–µ–≤—Å–∫–∏–π –ê–ª–µ–∫—Å–∞–Ω–¥—Ä –§–ª–æ—Ä–∏–∞–Ω –ñ–æ–∑–µ—Ñ (–Ω–µ–¥–æ–ø—É—Å–∫)
–¶–µ–∑–∞—Ä—å –ì–∞–π –Æ–ª–∏–π 4 (—ç–∫–∑–∞–º–µ–Ω)

–ü–æ—Å–ª–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ (–≤ —Å–∫–æ–±–∫–∞—Ö) ‚Äî —Å—Ç–∞—Ç—É—Å:
    ‚Ä¢ "—ç–∫–∑–∞–º–µ–Ω" ‚Äî —á–µ–ª–æ–≤–µ–∫ –ø—Ä–∏—à—ë–ª –Ω–∞ —ç–∫–∑–∞–º–µ–Ω –∏ –ø–æ–ª—É—á–∏–ª —Ç–∞–º –æ—Ü–µ–Ω–∫—É
    ‚Ä¢ "–∞–≤—Ç–æ–º–∞—Ç" ‚Äî —á–µ–ª–æ–≤–µ–∫ —Ö–æ—Ä–æ—à–æ —É—á–∏–ª—Å—è –∏ –ø–æ–ª—É—á–∏–ª –æ—Ü–µ–Ω–∫—É "–∞–≤—Ç–æ–º–∞—Ç–æ–º" –¥–æ —ç–∫–∑–∞–º–µ–Ω–∞
    ‚Ä¢ "–Ω–µ—è–≤–∫–∞" ‚Äî —á–µ–ª–æ–≤–µ–∫ –Ω–µ –ø—Ä–∏—à—ë–ª –Ω–∞ —ç–∫–∑–∞–º–µ–Ω –∏ —É –Ω–µ–≥–æ –Ω–µ—Ç –æ—Ü–µ–Ω–∫–∏
    ‚Ä¢ "–Ω–µ–¥–æ–ø—É—Å–∫" ‚Äî —á–µ–ª–æ–≤–µ–∫ –Ω–µ —Å–¥–∞–ª –∑–∞—á—ë—Ç –ø–æ —Ñ–∏–∑–∫—É–ª—å—Ç—É—Ä–µ, –∞ –ø–æ—Ç–æ–º—É –µ–≥–æ –Ω–µ –¥–æ–ø—É—Å—Ç–∏–ª–∏ –∫ —ç–∫–∑–∞–º–µ–Ω—É

–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –¢–û–õ–¨–ö–û –¥–ª—è —Ç–µ—Ö, –∫—Ç–æ —Å–¥–∞–≤–∞–ª —ç–∫–∑–∞–º–µ–Ω –æ—á–Ω–æ –∏–ª–∏ –ø–æ–ª—É—á–∏–ª –∞–≤—Ç–æ–º–∞—Ç.

–í–∞—à–∞ –∑–∞–¥–∞—á–∞:
1. –°—á–∏—Ç–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –ø–æ –≤—Å–µ–º —Å—Ç—É–¥–µ–Ω—Ç–∞–º –∏–∑ —Ñ–∞–π–ª–∞ `sheet`.
2. –í—ã—á–∏—Å–ª–∏—Ç—å —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª –ø–æ –≥—Ä—É–ø–ø–µ (—É—á–∏—Ç—ã–≤–∞—è —Ç–æ–ª—å–∫–æ —Å—Ç–∞—Ç—É—Å—ã "—ç–∫–∑–∞–º–µ–Ω" –∏ "–∞–≤—Ç–æ–º–∞—Ç").
3. –°—Ä–∞–≤–Ω–∏—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–Ω—ã–π —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª —Å —Ç–µ–º, —á—Ç–æ —É–∫–∞–∑–∞–Ω –ø—Ä–µ–ø–æ–¥–∞–≤–∞—Ç–µ–ª–µ–º –≤ —Ñ–∞–π–ª–µ `mean` (–≤ —Ñ–∞–π–ª–µ –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞ ‚Äî —á–∏—Å–ª–æ —Å –ø–ª–∞–≤–∞—é—â–µ–π —Ç–æ—á–∫–æ–π).
4. –ï—Å–ª–∏ —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª –≤–µ—Ä–µ–Ω ‚Äî –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å "OK"
   –ï—Å–ª–∏ —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª —Ä–∞—Å—Å—á–∏—Ç–∞–Ω —Å –æ—à–∏–±–∫–æ–π ‚Äî –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å "ERROR"

–ü—Ä–∏–º–µ—Ä—ã –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
    Sample Input 1:
        sheet = "sheet1.txt"
        mean = "mean1.txt"
    Sample Output 1:
        OK

    Sample Input 2:
        sheet = "sheet2.txt"
        mean = "mean1.txt"
    Sample Output 2:
        ERROR

    Sample Input 3:
        sheet = "sheet1.txt"
        mean = "mean2.txt"
    Sample Output 3:
        ERROR

    Sample Input 4:
        sheet = "sheet2.txt"
        mean = "mean2.txt"
    Sample Output 4:
        OK
"""

a = str(input())

f = open(a, 'r', encoding = 'utf-8')
t = f.read().splitlines()
k = 0
j = 0
for i in t:
    i = str(i)
    sh = i.split()
    # print(sh[-1], sh[-2])
    if sh[-1] == '(—ç–∫–∑–∞–º–µ–Ω)' or sh[-2] == '(–∞–≤—Ç–æ–º–∞—Ç)':
        k += 1
        j += int(sh[-2])
my_mean = str(int(round((j/k), 0)))
# print(my_mean)
f.close()

b = str(input())

f2 = open(b, 'r', encoding = 'utf-8')
txt = str(f2.read())
f2.close()
# print(txt)
if my_mean == txt:
    print('OK')
else:
    print('ERROR')
```
