```python
# –®–∞–≥ 2_8_2

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –ß—Ç–µ–Ω–∏–µ –∏–∑ —Ñ–∞–π–ª–∞

# –ü–æ—Å–ª–µ –æ—Ç–∫—Ä—ã—Ç–∏—è —Ñ–∞–π–ª–∞ –Ω–∞ —á—Ç–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, `open('file.txt', 'r')`) –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç—Ä–∏ –æ—Å–Ω–æ–≤–Ω—ã—Ö –º–µ—Ç–æ–¥–∞ –¥–ª—è —á—Ç–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–æ–≤–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏:

# 1. `read()` ‚Äî —á–∏—Ç–∞–µ—Ç **–≤–µ—Å—å —Ñ–∞–π–ª —Ü–µ–ª–∏–∫–æ–º** –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É.
# 2. `readline()` ‚Äî —á–∏—Ç–∞–µ—Ç **–æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É** (–≤–∫–ª—é—á–∞—è —Å–∏–º–≤–æ–ª –ø–µ—Ä–µ–Ω–æ—Å–∞ `\n`).
# 3. `readlines()` ‚Äî —á–∏—Ç–∞–µ—Ç **–≤–µ—Å—å —Ñ–∞–π–ª** –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ (–∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è `\n`).

# ‚ö†Ô∏è –í–∞–∂–Ω–æ:
# –í—Å–µ —ç—Ç–∏ –º–µ—Ç–æ–¥—ã –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç —Å—Ç—Ä–æ–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ **–≤–∫–ª—é—á–∞—é—Ç —Å–∏–º–≤–æ–ª –ø–µ—Ä–µ–Ω–æ—Å–∞ —Å—Ç—Ä–æ–∫–∏ `\n`** (–∫—Ä–æ–º–µ –ø–æ—Å–ª–µ–¥–Ω–µ–π —Å—Ç—Ä–æ–∫–∏, –µ—Å–ª–∏ –æ–Ω–∞ –Ω–µ –æ–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –ø–µ—Ä–µ–≤–æ–¥–æ–º —Å—Ç—Ä–æ–∫–∏).

# --- –ü—Ä–∏–º–µ—Ä—ã ---

# –ß–∏—Ç–∞–µ–º –≤–µ—Å—å —Ñ–∞–π–ª —Ü–µ–ª–∏–∫–æ–º:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
text = f.read()
print(text)  # ‚Üí –≤—ã–≤–æ–¥–∏—Ç –≤—Å—ë —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ñ–∞–π–ª–∞
f.close()

# –ß–∏—Ç–∞–µ–º –ø–æ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
first_line = f.readline()  # ‚Üí 'IE_NAME;IE_ID\n'
second_line = f.readline() # ‚Üí '–û–û–û ¬´–°–†-–ì—Ä—É–ø–ø¬ª;718436\n'
f.close()

# –ß–∏—Ç–∞–µ–º –≤—Å–µ —Å—Ç—Ä–æ–∫–∏ –≤ —Å–ø–∏—Å–æ–∫:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
lines = f.readlines()  # ‚Üí —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫, –∫–∞–∂–¥–∞—è —Å \n
f.close()

# --- –ü–æ—Å—Ç—Ä–æ—á–Ω–æ–µ —á—Ç–µ–Ω–∏–µ –≤ —Ü–∏–∫–ª–µ for ---

# Python –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å —Å—Ç—Ä–æ–∫–∏ —Ñ–∞–π–ª–∞ –Ω–∞–ø—Ä—è–º—É—é, –±–µ–∑ —Å–æ–∑–¥–∞–Ω–∏—è —Å–ø–∏—Å–∫–∞:

f = open('export_file_48603.csv', 'r', encoding='utf-8')
for line in f:
    print(line, end='')  # end='' —á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –¥–≤–æ–π–Ω–æ–≥–æ –ø–µ—Ä–µ–≤–æ–¥–∞ —Å—Ç—Ä–æ–∫–∏
f.close()

# --- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–µ –º–µ—Ç–æ–¥—ã —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ ---

# –ï—Å–ª–∏ –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ **–±–µ–∑ —Å–∏–º–≤–æ–ª–æ–≤ `\n`**, –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å:

# –°–ø–æ—Å–æ–± 1: split()
f = open('export_file_48603.csv', 'r', encoding='utf-8')
content = f.read()
strings = content.split('\n')  # ‚Üí —Ä–∞–∑–¥–µ–ª—è–µ–º –ø–æ —Å–∏–º–≤–æ–ª—É –ø–µ—Ä–µ–Ω–æ—Å–∞
f.close()

# –°–ø–æ—Å–æ–± 2 (—ç–ª–µ–≥–∞–Ω—Ç–Ω–µ–µ): splitlines()
f = open('export_file_48603.csv', 'r', encoding='utf-8')
strings = f.read().splitlines()  # ‚Üí —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ –±–µ–∑ \n
f.close()

# ‚ö†Ô∏è –í–Ω–∏–º–∞–Ω–∏–µ: –µ—Å–ª–∏ –≤ —Ñ–∞–π–ª–µ –µ—Å—Ç—å –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏ –∏–ª–∏ –¥–≤–æ–π–Ω—ã–µ –ø–µ—Ä–µ–Ω–æ—Å—ã, `splitlines()` –º–æ–∂–µ—Ç –≤–µ—Ä–Ω—É—Ç—å –ø—É—Å—Ç—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã ‚Äî –∏—Ö –∏–Ω–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ.

# üí° –°–æ–≤–µ—Ç: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `with` –¥–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–∞–∫—Ä—ã—Ç–∏—è —Ñ–∞–π–ª–∞!
```

```python
# –®–∞–≥ 2_8_4

# –ó–∞–¥–∞–Ω–∏–µ: –°—É–º–º–∞ —á–∏—Å–µ–ª –∏–∑ —Ñ–∞–π–ª–∞

# –°—á–∏—Ç–∞–π—Ç–µ –∏–º—è —Ñ–∞–π–ª–∞ —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã.
# –û—Ç–∫—Ä–æ–π—Ç–µ —Ñ–∞–π–ª –∏ –ø—Ä–æ—á–∏—Ç–∞–π—Ç–µ –µ–≥–æ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ.
# –§–∞–π–ª —Å–æ–¥–µ—Ä–∂–∏—Ç –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –∑–∞–ø–∏—Å–∞–Ω–æ —Å –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏.
# –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ –ø–µ—á–∞—Ç—å —Å—É–º–º—É –≤—Å–µ—Ö —ç—Ç–∏—Ö —á–∏—Å–µ–ª.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:
#   ‚Ä¢ –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–¥–∏—Ä–æ–≤–∫—É 'utf-8'.
#   ‚Ä¢ –ù–µ –∑–∞–±—É–¥—å—Ç–µ –∑–∞–∫—Ä—ã—Ç—å —Ñ–∞–π–ª –∏–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `with`.
#   ‚Ä¢ –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞–π—Ç–µ —Å—Ç—Ä–æ–∫–∏ –≤ —á–∏—Å–ª–∞ —Å –ø–æ–º–æ—â—å—é `int()`.

# –ü—Ä–∏–º–µ—Ä—ã:

# –í–≤–æ–¥: 1.txt ‚Üí —Å–æ–¥–µ—Ä–∂–∏—Ç: 1\n2 ‚Üí –≤—ã–≤–æ–¥: 3
# –í–≤–æ–¥: 2.txt ‚Üí —Å–æ–¥–µ—Ä–∂–∏—Ç: 1\n2\n3\n4 ‚Üí –≤—ã–≤–æ–¥: 10
# –í–≤–æ–¥: 3.txt ‚Üí —Å–æ–¥–µ—Ä–∂–∏—Ç: 1\n2\n...n (—Å—É–º–º–∞ –ø–µ—Ä–≤—ã—Ö 9 —á–∏—Å–µ–ª) ‚Üí –≤—ã–≤–æ–¥: 45

a = str(input())
f = open(a, 'r', encoding = 'utf-8')
# c = f.read().split('\n')
c = f.read().splitlines()
sum = 0
for k in c:
    sum += int(k)
print(sum)
f.close()
```
