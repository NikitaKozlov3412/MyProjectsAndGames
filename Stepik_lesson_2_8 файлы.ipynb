```python
# –®–∞–≥ 2_8_2

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –ß—Ç–µ–Ω–∏–µ –∏–∑ —Ñ–∞–π–ª–∞

# –ü–æ—Å–ª–µ –æ—Ç–∫—Ä—ã—Ç–∏—è —Ñ–∞–π–ª–∞ –Ω–∞ —á—Ç–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, `open('file.txt', 'r')`) –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç—Ä–∏ –æ—Å–Ω–æ–≤–Ω—ã—Ö –º–µ—Ç–æ–¥–∞ –¥–ª—è —á—Ç–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–æ–≤–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏:

# 1. `read()` ‚Äî —á–∏—Ç–∞–µ—Ç **–≤–µ—Å—å —Ñ–∞–π–ª —Ü–µ–ª–∏–∫–æ–º** –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É.
# 2. `readline()` ‚Äî —á–∏—Ç–∞–µ—Ç **–æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É** (–≤–∫–ª—é—á–∞—è —Å–∏–º–≤–æ–ª –ø–µ—Ä–µ–Ω–æ—Å–∞ `\n`).
# 3. `readlines()` ‚Äî —á–∏—Ç–∞–µ—Ç **–≤–µ—Å—å —Ñ–∞–π–ª** –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ (–∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è `\n`).

# ‚ö†Ô∏è –í–∞–∂–Ω–æ:
# –í—Å–µ —ç—Ç–∏ –º–µ—Ç–æ–¥—ã –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç —Å—Ç—Ä–æ–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ **–≤–∫–ª—é—á–∞—é—Ç —Å–∏–º–≤–æ–ª –ø–µ—Ä–µ–Ω–æ—Å–∞ —Å—Ç—Ä–æ–∫–∏ `\n`** (–∫—Ä–æ–º–µ –ø–æ—Å–ª–µ–¥–Ω–µ–π —Å—Ç—Ä–æ–∫–∏, –µ—Å–ª–∏ –æ–Ω–∞ –Ω–µ –æ–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –ø–µ—Ä–µ–≤–æ–¥–æ–º —Å—Ç—Ä–æ–∫–∏).

# --- –ü—Ä–∏–º–µ—Ä—ã ---

# –ß–∏—Ç–∞–µ–º –≤–µ—Å—å —Ñ–∞–π–ª —Ü–µ–ª–∏–∫–æ–º:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
text = f.read()
print(text)  # ‚Üí –≤—ã–≤–æ–¥–∏—Ç –≤—Å—ë —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ñ–∞–π–ª–∞
f.close()

# –ß–∏—Ç–∞–µ–º –ø–æ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
first_line = f.readline()  # ‚Üí 'IE_NAME;IE_ID\n'
second_line = f.readline() # ‚Üí '–û–û–û ¬´–°–†-–ì—Ä—É–ø–ø¬ª;718436\n'
f.close()

# –ß–∏—Ç–∞–µ–º –≤—Å–µ —Å—Ç—Ä–æ–∫–∏ –≤ —Å–ø–∏—Å–æ–∫:
f = open('export_file_48603.csv', 'r', encoding='utf-8')
lines = f.readlines()  # ‚Üí —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫, –∫–∞–∂–¥–∞—è —Å \n
f.close()

# --- –ü–æ—Å—Ç—Ä–æ—á–Ω–æ–µ —á—Ç–µ–Ω–∏–µ –≤ —Ü–∏–∫–ª–µ for ---

# Python –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å —Å—Ç—Ä–æ–∫–∏ —Ñ–∞–π–ª–∞ –Ω–∞–ø—Ä—è–º—É—é, –±–µ–∑ —Å–æ–∑–¥–∞–Ω–∏—è —Å–ø–∏—Å–∫–∞:

f = open('export_file_48603.csv', 'r', encoding='utf-8')
for line in f:
    print(line, end='')  # end='' —á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –¥–≤–æ–π–Ω–æ–≥–æ –ø–µ—Ä–µ–≤–æ–¥–∞ —Å—Ç—Ä–æ–∫–∏
f.close()

# --- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–µ –º–µ—Ç–æ–¥—ã —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ ---

# –ï—Å–ª–∏ –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ **–±–µ–∑ —Å–∏–º–≤–æ–ª–æ–≤ `\n`**, –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å:

# –°–ø–æ—Å–æ–± 1: split()
f = open('export_file_48603.csv', 'r', encoding='utf-8')
content = f.read()
strings = content.split('\n')  # ‚Üí —Ä–∞–∑–¥–µ–ª—è–µ–º –ø–æ —Å–∏–º–≤–æ–ª—É –ø–µ—Ä–µ–Ω–æ—Å–∞
f.close()

# –°–ø–æ—Å–æ–± 2 (—ç–ª–µ–≥–∞–Ω—Ç–Ω–µ–µ): splitlines()
f = open('export_file_48603.csv', 'r', encoding='utf-8')
strings = f.read().splitlines()  # ‚Üí —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ –±–µ–∑ \n
f.close()

# ‚ö†Ô∏è –í–Ω–∏–º–∞–Ω–∏–µ: –µ—Å–ª–∏ –≤ —Ñ–∞–π–ª–µ –µ—Å—Ç—å –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏ –∏–ª–∏ –¥–≤–æ–π–Ω—ã–µ –ø–µ—Ä–µ–Ω–æ—Å—ã, `splitlines()` –º–æ–∂–µ—Ç –≤–µ—Ä–Ω—É—Ç—å –ø—É—Å—Ç—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã ‚Äî –∏—Ö –∏–Ω–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ.

# üí° –°–æ–≤–µ—Ç: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `with` –¥–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–∞–∫—Ä—ã—Ç–∏—è —Ñ–∞–π–ª–∞!
```

```python
# –ü–æ–ª–µ–∑–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è

import os

# 1. –ß–∏—Ç–∞–µ–º –ø—É—Ç—å (–æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –∏–ª–∏ –∞–±—Å–æ–ª—é—Ç–Ω—ã–π) –∏–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –≤–≤–æ–¥–∞
s = input()

# 2. –ü—Ä–æ–≤–µ—Ä—è–µ–º: —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –ø—É—Ç—å —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–º –§–ê–ô–õ–û–ú?
if os.path.isfile(s):
    # ‚úÖ –§–∞–π–ª —Å—É—â–µ—Å—Ç–≤—É–µ—Ç ‚Äî –±–µ–∑–æ–ø–∞—Å–Ω–æ –æ—Ç–∫—Ä—ã–≤–∞–µ–º –µ–≥–æ —Å –ø–æ–º–æ—â—å—é –º–µ–Ω–µ–¥–∂–µ—Ä–∞ –∫–æ–Ω—Ç–µ–∫—Å—Ç–∞
    with open(s, 'r', encoding='utf-8') as f:
        print('CONTENT:')
        # –ò—Ç–µ—Ä–∏—Ä—É–µ–º—Å—è –ø–æ —Å—Ç—Ä–æ–∫–∞–º —Ñ–∞–π–ª–∞ (–ø–æ—Å—Ç—Ä–æ—á–Ω–æ–µ —á—Ç–µ–Ω–∏–µ ‚Äî —ç–∫–æ–Ω–æ–º–∏—Ç –ø–∞–º—è—Ç—å)
        for line in f:
            # .strip() —É–¥–∞–ª—è–µ—Ç —Å–∏–º–≤–æ–ª—ã –ø–µ—Ä–µ–≤–æ–¥–∞ —Å—Ç—Ä–æ–∫–∏ –∏ –ø—Ä–æ–±–µ–ª—ã –ø–æ –∫—Ä–∞—è–º
            # ‚ö†Ô∏è –í–∞–∂–Ω–æ: –µ—Å–ª–∏ –≤ —Ñ–∞–π–ª–µ –µ—Å—Ç—å –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏, –æ–Ω–∏ —Å—Ç–∞–Ω—É—Ç '' –∏ –Ω–µ –æ—Ç–æ–±—Ä–∞–∑—è—Ç—Å—è
            # –ü—Ä–∏ –∂–µ–ª–∞–Ω–∏–∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å —Å—Ç—Ä—É–∫—Ç—É—Ä—É (–≤–∫–ª—é—á–∞—è –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏), –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å .rstrip('\n')
            print(line.strip())

# 3. –ï—Å–ª–∏ –ø—É—Ç—å ‚Äî —ç—Ç–æ –°–£–©–ï–°–¢–í–£–Æ–©–ò–ô –ö–ê–¢–ê–õ–û–ì:
elif os.path.isdir(s):
    print('ERROR:')
    print('–≠—Ç–æ –∫–∞—Ç–∞–ª–æ–≥, –∞ –Ω–µ —Ñ–∞–π–ª')
    # ‚úÖ –õ—É—á—à–µ –ø–µ—á–∞—Ç–∞—Ç—å —á–µ—Ä–µ–∑ –¥–≤–∞ print ‚Äî —Ç–∞–∫ —Ç–æ—á–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—Å—è –ø–µ—Ä–µ–Ω–æ—Å —Å—Ç—Ä–æ–∫–∏.
    # –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞: print('ERROR:\n–≠—Ç–æ –∫–∞—Ç–∞–ª–æ–≥, –∞ –Ω–µ —Ñ–∞–π–ª')

# 4. –õ—é–±–æ–π –¥—Ä—É–≥–æ–π —Å–ª—É—á–∞–π: —Ñ–∞–π–ª–∞/–∫–∞—Ç–∞–ª–æ–≥–∞ –Ω–µ—Ç, –∏–ª–∏ –ø—É—Ç—å –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π:
else:
    print('ERROR:')
    print('–§–∞–π–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç')
```

```python
# –®–∞–≥ 2_8_12

# –ó–∞–¥–∞–Ω–∏–µ: –ü—Ä–æ–≤–µ—Ä–∫–∞ –∏ —á—Ç–µ–Ω–∏–µ —Ñ–∞–π–ª–∞ –ø–æ –ø–æ–ª–Ω–æ–º—É –ø—É—Ç–∏

# –ù–∞ –≤—Ö–æ–¥ –ø–æ–¥–∞—ë—Ç—Å—è –ø–æ–ª–Ω—ã–π –ø—É—Ç—å –∫ —Ñ–∞–π–ª—É –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Ç–µ–∫—É—â–µ–≥–æ –∫–∞—Ç–∞–ª–æ–≥–∞.

# –í–∞—à–∞ –∑–∞–¥–∞—á–∞:
# 1. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ –ø–æ —É–∫–∞–∑–∞–Ω–Ω–æ–º—É –ø—É—Ç–∏ **—Ñ–∞–π–ª** (–Ω–µ –∫–∞—Ç–∞–ª–æ–≥).
# 2. –ï—Å–ª–∏ —ç—Ç–æ —Ñ–∞–π–ª ‚Äî –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ –ø–µ—á–∞—Ç—å:
#      "CONTENT:"
#      –∑–∞—Ç–µ–º —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ñ–∞–π–ª–∞ (–≤—Å—ë, —á—Ç–æ –≤ –Ω—ë–º –µ—Å—Ç—å).
# 3. –ï—Å–ª–∏ —ç—Ç–æ **–∫–∞—Ç–∞–ª–æ–≥** ‚Äî –≤—ã–≤–µ–¥–∏—Ç–µ:
#      "ERROR:"
#      "–≠—Ç–æ –∫–∞—Ç–∞–ª–æ–≥, –∞ –Ω–µ —Ñ–∞–π–ª"
# 4. –ï—Å–ª–∏ —Ñ–∞–π–ª–∞ –∏–ª–∏ –∫–∞—Ç–∞–ª–æ–≥–∞ **–Ω–µ—Ç** ‚Äî –≤—ã–≤–µ–¥–∏—Ç–µ:
#      "ERROR:"
#      "–§–∞–π–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç"

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:
#   ‚Ä¢ –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ `os.path.isfile()` –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏, —á—Ç–æ –ø—É—Ç—å —É–∫–∞–∑—ã–≤–∞–µ—Ç –∏–º–µ–Ω–Ω–æ –Ω–∞ —Ñ–∞–π–ª.
#   ‚Ä¢ –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ `with open(...)` –¥–ª—è –±–µ–∑–æ–ø–∞—Å–Ω–æ–≥–æ —á—Ç–µ–Ω–∏—è.
#   ‚Ä¢ –ù–µ –∑–∞–±—É–¥—å—Ç–µ —É–∫–∞–∑–∞—Ç—å –∫–æ–¥–∏—Ä–æ–≤–∫—É 'utf-8'.

# –ü—Ä–∏–º–µ—Ä—ã:

# –í–≤–æ–¥: tmp/dir/text.txt ‚Üí –≤—ã–≤–æ–¥: CONTENT:\n–ü—Ä–∏–≤–µ—Ç, –ú–∏—Ä!
# –í–≤–æ–¥: tmp/dir/        ‚Üí –≤—ã–≤–æ–¥: ERROR:\n–≠—Ç–æ –∫–∞—Ç–∞–ª–æ–≥, –∞ –Ω–µ —Ñ–∞–π–ª
# –í–≤–æ–¥: tmp/text.txt     ‚Üí –≤—ã–≤–æ–¥: ERROR:\n–§–∞–π–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç

import os # –∏–ª–∏ import os.path

a = input()

if os.path.isfile(a):
    with open(a, 'r', encoding='utf-8') as f:
        print('CONTENT:')
        for str in f:
            print(str.strip())
elif os.path.isdir(a):
    print('ERROR:')
    print('–≠—Ç–æ –∫–∞—Ç–∞–ª–æ–≥, –∞ –Ω–µ —Ñ–∞–π–ª')
else:
    print('ERROR:')
    print('–§–∞–π–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç')
```
