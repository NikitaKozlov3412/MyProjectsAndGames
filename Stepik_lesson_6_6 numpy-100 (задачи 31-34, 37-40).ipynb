```python
# Шаг 6_6_7

# 37.1

# На вход подаётся 2 строки:
#   - 1-я строка содержит 2 натуральных числа: n, m
#   - 2-я строка содержит число k

# Создайте матрицу размера n*m такую, что каждая её строка содержит числа от k до k+m-1 (с шагом 1).

# Ответ запишите в переменную Z.

import numpy as np

n,m = list(map(int, input().split()))
k = int(input())
# print(n,m,k)

Z = np.tile(np.arange(k, k+m),(n,1)).astype(float)
Z
```

```python
# Шаг 6_6_8

# 37.2

# На вход подаётся 2 строки:
#   - 1-я строка содержит 2 натуральных числа: n, m
#   - 2-я строка содержит число k

# Создайте матрицу размера n*m такую, что каждый её столбец содержит числа от k до k+n-1 (с шагом 1).

# Ответ запишите в переменную Z.

import numpy as np

n,m = list(map(int, input().split()))
k = int(input())
# print(n,m,k)

Z = np.tile(np.arange(k, k+n),(m,1)).astype(float).transpose()
Z
```

```python
# Решение через reshape:

import numpy as np

n,m = list(map(int, input().split()))
k = int(input())
# print(n,m,k)

Z = np.full((n,m),k, dtype=float) + np.arange(n).reshape(n,1)
Z
```

```python
# Шаг 6_6_10

# 39.1 (деление отрезка на N+1 часть)

# На вход подаются 3 числа (каждое с новой строки):
#   - start
#   - stop
#   - n

# Составьте список (numpy.array) из n точек на отрезке (start, stop), попарно равноудалённых друг от друга и от концов отрезка (т.е. разделите отрезок на n+1 часть).

# Округлите значения точек до 3 знака после запятой.

# Результат сохраните в переменную Z.

import numpy as np

start = int(input())
stop = int(input())
n = int(input())

Z = np.round(np.linspace(start, stop, n+1, endpoint = False)[1:], 3)

print(Z)

# Дока: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
```

```python
# Шаг 6_6_11

# 39.2

# На вход подаются 3 числа (каждое с новой строки):
#   - start
#   - stop
#   - n

# Составьте список из n точек на отрезке [start, stop] в геометрической прогрессии, включая start и stop.

# Округлите значения точек до 3 знака после запятой.

# Результат сохраните в переменную Z.

import numpy as np

start = int(input())
stop = int(input())
n = int(input())

Z = np.geomspace(start, stop, n).round(3)

print(Z)

# Дока: https://numpy.org/doc/stable/reference/generated/numpy.geomspace.html#numpy.geomspace
```

```python
# Шаг 6_6_12

# 40

# Считайте 2 числа:
#   - seed
#   - n

# С помощью Numpy сгенерируйте n случайных чисел из интервала (0, 1) с фиксированным seed.
# Отсортируйте их по возрастанию и сохраните в переменную Z.

# Дока : https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.random.seed.html
# Дока: https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.random.html

import numpy as np

seed = int(input())
n = int(input())

np.random.seed(seed)

Z = np.sort(np.random.random_sample((n,)))
print(Z)
```
