```python
# Шаг 4_5_1

# Тема: Таблица умножения двузначных чисел от "Мудреца"

# Рассмотрим схему "упрощённого" умножения больших двузначных чисел.

# Схема:
#   X · Y = [ (100 - ((100-X) + (100-Y)) ) · 100 ] + [ (100-X) · (100-Y) ]

# Где:
#   • Первая часть: первые 2 цифры результата.
#   • Вторая часть: последние 2 цифры результата.

# По шагам:
#   1. Вычитаем из 100 оба числа → получаем (100-X) и (100-Y).
#   2. Складываем (100-X) и (100-Y).
#   3. Вычитаем сумму из 100.
#   4. Умножаем (100-X) и (100-Y).
#   5. Первые 2 цифры — результат шага 3, умноженный на 100.
#      Последние 2 цифры — результат шага 4.

# Почему первый слагаемый умножается на 100?
#   Чтобы он не влиял на последние 2 цифры результата.

# Математически это тождественное равенство:
#   (100 - ((100-x) + (100-y))) * 100 + (100-x)*(100-y) = x*y

# Однако в этой схеме есть несколько "хитростей", которые мы разберём в следующих задачах.

# Примечание:
#   Не все задачи этого раздела можно оценить программно — вы можете решать их с помощью TeX или иных уловок.
#   Такие задачи не приносят баллы, но постарайтесь решить их честно.
```

```python
# Шаг 4_5_2

# Задание: simple_multiplication(x, y)

# Напишите функцию `simple_multiplication(x, y)`, реализующую умножение по схеме "Мудреца".

# Функция должна возвращать целое число — результат умножения.

# Примеры:
#   simple_multiplication(96, 97) → 9312
#   simple_multiplication(84, 16) → 1344
#   simple_multiplication(10, 10) → 100

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

def simple_multiplication(x, y):
    x = int(x)
    y = int(y)
    a = 100 - x
    b = 100 - y
    tot1 = (100 - a - b) * 100 + a * b
    return tot1

simple_multiplication(10, 10)

```

```python
# Шаг 4_5_6

# Тема: Техническое решение "схемы Мудреца" — конкатенация строк

# Надо понимать, что идея "схемы Мудреца" — это упрощение математических операций, их механизация.
# По сути, мы хотим получить 2 числа и просто записать их друг за другом — а не проводить сложение.

# Таким образом, "схема Мудреца" из 1 иллюстрации предполагает скорее **конкатенацию** двух строк.

# Пример: 99×91

# По шагам:
#   1. Вычитаем из 100 оба числа → (100-99=1) и (100-91=9).
#   2. Складываем результаты шага 1 → 1+9 = 10.
#   3. Вычитаем из 100 результат сложения → 100-10 = 90.
#   4. Перемножаем результаты шага 1 → 1×9 = 9.
#   5. Результат шага 3 даёт первые 2 цифры результата, а результат шага 4 — последние 2 цифры.

# ПОГОДИТЕ! Результат шага 4 — однозначное число: 9.
# Как нам с его помощью записать 2 последних цифры?

# Необходимо дописать 0!

# Итоговый результат:
#   Шаг 3: 90 → первые 2 цифры → "90"
#   Шаг 4: 9 → последние 2 цифры → "09" (дописали ноль спереди)
#   Конкатенация: "90" + "09" = "9009"

# Это техническое решение — оно может иметь ещё одно последствие:
#   Всегда использовать строки и форматировать вторую часть как 2-значную (с ведущими нулями).

# Такой подход позволяет избежать математики и работать со строками — что часто проще для программирования.
```

```python
# Шаг 4_5_7

# Задание: wisdom_multiplication(x, y, length_check=True)

# Напишите функцию `wisdom_multiplication(x, y, length_check=True)`, реализующую умножение по схеме "Мудреца".

# По шагам:
#   1. Вычитаем из 100 оба числа → a = 100 - x, b = 100 - y.
#   2. Складываем результаты шага 1 → s = a + b.
#   3. Вычитаем из 100 результат сложения → first_part = 100 - s.
#   4. Перемножаем результаты шага 1 → second_part = a * b.
#   5. Результат шага 3 даёт первые цифры результата, а результат шага 4 — последние цифры.

# В зависимости от значения аргумента `length_check`:
#   • Если `True` — функция проверяет длину `second_part`.
#     - Если `second_part` состоит из 1 цифры — дописывает ведущий ноль (например, 9 → "09").
#     - Если `second_part` состоит из 2 цифр — оставляет как есть.
#   • Если `False` — не проверяет длину и использует `second_part` как есть.

# Функция должна возвращать целое число — результат умножения.

# Примеры:
#   wisdom_multiplication(96, 97) → 9312
#   wisdom_multiplication(99, 99, length_check=True) → 9801
#   wisdom_multiplication(99, 99, length_check=False) → 981
#   wisdom_multiplication(10, 10) → -808100 (без length_check)
#   wisdom_multiplication(91, 99) → 9009

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

def wisdom_multiplication(x, y, length_check = True):
    a = 100 - int(x)
    b = 100 - int(y)
    c = a + b
    d = 100 - c
    e = a * b
    if length_check is True:
        if len(str(e)) < 2:
            e = "0" + str(e)
    tot = str(d) + str(e)
    return int(tot)

wisdom_multiplication(91, 99)

```
