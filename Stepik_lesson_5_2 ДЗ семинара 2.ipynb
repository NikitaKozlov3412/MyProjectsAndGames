```python
# Шаг 5_2_3

# 2. Производная

# Напишите функцию derivative(f, x0=0), находящую производную функции f(x) в точке x = x0

# Функция должна принимать 2 параметра:
#   - обязательный - f - имя функции (без скобочек)
#   - необязательный - x0 - значение аргумента точки в которой вычисляется производная.
#     Если не задан, то значение будет равно 0.

# Функция должна округлять ответ до 3 знака после запятой.

# Примечание. В этой задаче не нужно ничего считывать и ничего выводить на печать.
# Только реализовать функцию.

# Sample Input 1:
#   derivative(sin)

# Sample Output 1:
#   1.0

# Sample Input 2:
#   derivative(cos)

# Sample Output 2:
#   0.0

# Sample Input 3:
#   derivative(atan, 1000000)

# Sample Output 3:
#   0.0

# Sample Input 4:
#   derivative(exp, 1)

# Sample Output 4:
#   2.718

import math

def derivative(f, x0=0):
    dx = 1e-7
    return round((f(x0 +dx) - f(x0)) / dx , 3)

derivative(math.exp, 1)
```

```python
# Шаг 5_2_4

# 3. Вытянуть многомерный список

# Напишите функцию list_pull(L), принимающую на вход многомерный список (список списков) вида:
#   [['one'], [343, 2], [[9, 9, 9], [[666, 666], [[[[42]]]]]]

# и возвращающую "вытянутый" список только значений вида:
#   ['one', 343, 2, 9, 9, 9, 666, 666, 42]

# Условия:
#   - Принимаемый список может иметь произвольное число дочерних списков
#   - Длины дочерних списков могут быть произвольными (дочерний список может не содержать элементов) и не равны друг другу
#   - Элементы дочерних списков могут повторяться
#   - Порядок в получившемся списке не важен

# Примечание1. В этой задаче не надо ничего считывать или выводить на печать.

# Примечание2. Тип переменной можно узнать с помощью type() и сравнить с типом list:
#   type(elem) is list  # Вернёт True для списков и False для всех остальных

# Это усложнённая версия задачи "Шаг 7 - Вытянуть двумерный список"


def parser(L):
    L_parse = []
    for L1 in L:
        if type(L1) is list:
            for L2 in L1:
                L_parse.append(L2)
        else:
            L_parse.append(L1)

    return L_parse

def list_pull(L):
    count = 0
    for L11 in L:
        if type(L11) is list:
            count += 1
    # if count > 0:
        # L = parser(L)
    while count > 0:
        count = 0
        L = parser(L)
        for L11 in L:
            if type(L11) is list:
                count += 1
    return L



list_pull([['one'], [343, 2], [[9, 9, 9], [[666, 666], [[[[42]]]]]]])
```

```python
# Шаг 5_2_5

# 4. Скопировать многомерный список без deepcopy

# На вход подаётся многоуровневый список L1 вида:
#   L1 = [['one'], [343, 2], [[9, 9, 9], [[666, 666], [[[[42]]]]]]

# Не используя метод deepcopy скопируйте всё содержимое списка L1 в L2.
# Учтите, что элементы дочерних элементов L1 после этого будут изменены и эти изменения не должны отразиться на L2.
# Т.е. сделайте по настоящему глубокую копию.

# Условия:
#   - Принимаемый список может иметь произвольное число дочерних списков
#   - Длины дочерних списков могут быть произвольными (дочерний список может не содержать элементов) и не равны друг другу
#   - Элементы дочерних списков могут повторяться
#   - Порядок в получившемся списке не важен

# Примечание1. В этой задаче не надо ничего считывать или выводить на печать.

# Примечание2. Тип переменной можно узнать с помощью type() и сравнить с типом list:
#   type(elem) is list  # Вернёт True для списков и False для всех остальных

# Примечание3. Deepcopy использовать нельзя

def surface_copy(input_list):
    if isinstance(input_list, list):
        return [surface_copy(list_L) for list_L in input_list]
    else:
        return input_list


def deep_copy(L1):
    L2 = surface_copy(L1)
    return L2

deep_copy([['one'], [343, 2], [[9, 9, 9], [[666, 666], [[[[42]]]]]]])
```

```python
# Шаг 5_2_6

# A. verbing

# Напишите функцию verbing(s), принимающую на вход строку.

# Условия:
#   - Если длина строки 3 и больше, то добавьте к ней 'ing' в конце.
#   - Если строка уже содержит 'ing', добавьте 'ly'.
#   - Если длина строки меньше 3, верните строку как есть.

# Примечание. В этой задаче не нужно ничего считывать и ничего выводить на печать. Только реализовать функцию.

# Sample Input 1:
#   hail

# Sample Output 1:
#   hailing

# Sample Input 2:
#   swimming

# Sample Output 2:
#   swimmingly

# Sample Input 3:
#   do

# Sample Output 3:
#   do

def verbing(L):
    L2 = str(L)
    if len(L2) < 3:
        return L2
    elif 'ing' in L2:
        L2 = L2 + 'ly'
        return L2
    elif len(L2) >=3 and 'ing' not in L2:
        L2 = L2 + 'ing'
        return L2

```

```python
# Шаг 5_2_7

# B. front_back

# Рассмотрим разделение строки на две половины.
#   - Если длина чётная, 1я и 2я части имеют одинаковую длину.
#   - Если длина нечётная, дополнительный символ идёт в передней половине.

# Например, «abcde»: 1я часть — «abc», 2я часть — «de».

# Напишите функцию front_back(a, b), принимающую 2 строки, a и b,
# и возвращающую строку формы:
#   # a-front + b-front + a-back + b-back

# Примечание. В этой задаче не нужно ничего считывать и ничего выводить на печать. Только реализовать функцию.

# Sample Input 1:
#   abcd
#   xy

# Sample Output 1:
#   abxcyd

# Sample Input 2:
#   abcde
#   xyz

# Sample Output 2:
#   abcxydez

# Sample Input 3:
#   Kitten
#   Donut

# Sample Output 3:
#   KitDontenut

def front_back(a, b):
    a = str(a)
    b = str(b)
    if len(a) % 2 == 0:
        c = len(a) // 2
        a_front = a[:c]
        a_back = a[c:]
    else:
        c = (len(a) // 2) + 1
        a_front = a[:c]
        a_back = a[c:]
    if len(b) % 2 == 0:
        d = len(b) // 2
        b_front = b[:d]
        b_back = b[d:]
    else:
        d = (len(b) // 2) + 1
        b_front = b[:d]
        b_back = b[d:]
    tot_str = a_front + b_front + a_back + b_back
    return tot_str

front_back('Kitten', 'Donut')
```

```python
# Хитрое решение из комментов

def front_back(a, b):
    sep = lambda x: (len(x) + 1) // 2
    return a[:sep(a)] + b[:sep(b)] + a[sep(a):len(a)] + b[sep(b):len(b)]

front_back('Kitten', 'Donut')
```

```python
# Шаг 5_2_8

# C1. mimic_dict

# Напишите функцию mimic_dict(string), которая принимает на вход строковую переменную (может содержать пробелы, табуляцию, переносы строк),
# возвращающую «имитирующий» словарь, который соотносит каждое появившееся слово, со списком всех слов, которые следуют за этим словом.

# Для "входа" в словарь используйте в качестве ключа пустую строку, в соответствие которой будет поставлено 1 слово.

# Примечание. В этой задаче не нужно ничего считывать и ничего выводить на печать. Только реализовать функцию.

# Sample Input 1:
#   Uno dos tres cuatro cinco

# Sample Output 1:
#   {"": ["Uno"], "Uno": ["dos"], "dos": ["tres"], "tres": ["cuatro"], "cuatro": ["cinco"]}

# Sample Input 2:
#   a cat and a dog
#   a fly

# Sample Output 2:
#   {"": ["a"], "a": ["cat", "dog", "fly"], "cat": ["and"], "and": ["a"], "dog": ["a"]}

def mimic_dict(string):
    dict = {}
    string = string.split()
    set_str = set(string)
    # print(string)
    # print(set_str)
    list_str = []
    for i in string:
        if i in set_str and i not in list_str:
            list_str.append(i)
    # print(list_str)
    dict[""] = [list_str[0]]
    # print(dict)
    for k in range(0, len(list_str) - 1):
        # print(list_str[k])
        value_p = []
        for j in range(len(string) - 1):
            if string[j] == list_str[k]:
                value_p.append(string[j+1])
        # print(value_p)
        dict[list_str[k]] = value_p
    return dict
```
