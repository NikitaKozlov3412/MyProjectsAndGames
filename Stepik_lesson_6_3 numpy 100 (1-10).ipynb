```python
# Шаг 6_3_7

# 3.2

# Считайте строку, где значения заданы через пробел: сначала shape, затем dtype.

# Последнее значение в строке может быть либо числом (тогда это последнее значение кортежа shape), либо строкой dtype.

#   - shape — атрибут, задающий размеры матрицы нулей (позволяет вернуть вектор-строку, вектор-столбец, матрицу, куб и т.д.)
#   - dtype — тип данных, использующихся для значений матрицы. По-умолчанию используется numpy.float64 (если dtype не указан).

# Создайте матрицу Z размера shape, со значениями типа dtype.

# Примечание. В этой задаче не нужно ничего выводить на печать. Только создать матрицу Z.

# Черновик:

str = input().split()

print(str)

str_0 = str[:-1]
str_1 = str[-1]
print(str_0)
print(str_1)
str_0 = list(map(int, str_0))
print(str_0)
if str_1.isdigit():
    str_0.append(int(str_1))
else:
    d_type = str_1
print(str_0)

str_0 = tuple(str_0)

Z = np.zeros((str_0), dtype = d_type)

print(Z)
```

```python
# Шаг 6_3_8

# 4

# Посчитайте размер матрицы Z в байтах и выведите его на печать.

# Чтобы узнать размер (в байтах) 1 элемента матрицы используйте атрибут https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.ndarray.itemsize.html

# Примечание. В этой задаче не нужно ничего считывать — матрица Z уже заполнена.

import numpy as np

Z = np.zeros((10,10))
k = list(Z.shape)
# print(k)
tot = Z.itemsize
for i in k:
    tot =  tot * i

print(tot)
```
