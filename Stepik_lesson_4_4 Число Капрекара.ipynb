```python
# Шаг 4_4_1

# Задание: kaprekar(n)

# Напишите функцию `kaprekar(n)`, принимающую на вход натуральное число и возвращающую:
#   • `True`, если число является Числом Капреакара,
#   • `False`, если число НЕ является Числом Капреакара.

# Определение:
#   Число n называется Числом Капреакара, если его квадрат можно разбить на две части (необязательно равные),
#   сумма которых даст исходное число n.
#   Вторая часть может начинаться с 0, но не должна быть пустой.

# Примеры:
#   9 → 9² = 81 → 8 + 1 = 9 → True
#   297 → 297² = 88209 → 88 + 209 = 297 → True
#   4879 → 4879² = 23804641 → 238 + 04641 = 4879 → True
#   100 → 100² = 10000 → 100 + 00 = 100? — нет, вторая часть пустая → False

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

n = int(input())

def kaprekar(n):
    a = int(n**2)
    g = str(a)

    for i in range(1, len(g)):
        c = int(g[:i])
        d = int(g[i:])
        if c + d == n and d != 0:
            return True
            break
        else:
            i += 1
    return False




kaprekar(n)
```
