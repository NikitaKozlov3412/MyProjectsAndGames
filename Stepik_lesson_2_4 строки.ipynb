```python
# Шаг 2_4_1

# --- Строки в Python ---

# Строки — это упорядоченные последовательности символов, используемые для хранения текста.

# Всё, что вводится через input() или читается из файла (open, read), — это строки.
a = 'string'       # строка в одинарных кавычках
b = "new string"   # строка в двойных кавычках

# Тройные кавычки — для многострочных строк
c = '''line 1
line 2
line 3'''
d = """New line 1
new line 2
new line 3"""

# Можно использовать разные кавычки внутри строки
S1 = 'spam"s'   # одинарные снаружи, двойные внутри
S2 = "spam's"   # двойные снаружи, одинарные внутри


# --- Экранирование символов ---

# Для экранирования используется символ \
s1 = 'some \'symbols\' inside'  # экранирование кавычек

# Спецсимволы:
# \n  — перевод строки
# \t  — табуляция
# \\  — обратный слеш
# \a  — звонок
# \b  — забой
# \r  — возврат каретки
# \f  — перевод страницы
# \v  — вертикальная табуляция
# \N{name} — символ Юникода по имени
# \xhh — символ в 16-ричном формате
# \uhhhh — 16-битный Юникод
# \Uhhhhhhhh — 32-битный Юникод
# \ooo — 8-ричное значение символа
# \0 — символ null


# --- Unicode ---

# В Python 2 нужно было явно указывать u перед строкой
# us = u'Unicode строка в Python 2'

# В Python 3 все строки по умолчанию являются Unicode
text = "Пример строки с Юникодом"
print(text)


```

```python
# --- Примеры экранирования символов в Python ---

print("1. Перевод строки (\\n):")
print("Первая строка\nВторая строка")

print("\n2. Горизонтальная табуляция (\\t):")
print("Слово1\tСлово2\tСлово3")

print("\n3. Обратный слеш (\\\\):")
print("Путь к файлу: C:\\Users\\User\\Documents")

print("\n4. Звонок (\\a):")
print("Сейчас будет звуковой сигнал!\a")  # может не сработать в некоторых IDE

print("\n5. Возврат каретки (\\r):")
print("12345\rabc")  # затирает начало строки

print("\n6. Перевод страницы (\\f):")
print("Первая страница\fВторая страница")  # эффект заметен не всегда

print("\n7. Вертикальная табуляция (\\v):")
print("Слово1\vСлово2\vСлово3")

print("\n8. Символ Unicode по имени (\\N{name}):")
print("\N{GREEK SMALL LETTER DELTA}")  # греческая дельта δ

print("\n9. Символ по 16-ричному коду (\\xhh):")
print("\x41\x42\x43")  # символы A B C

print("\n10. Символ по 16-битному коду Unicode (\\uhhhh):")
print("\u041F\u0440\u0438\u0432\u0435\u0442")  # «Привет» на русском

print("\n11. Символ по 32-битному коду Unicode (\\Uhhhhhhhh):")
print("\U0001F600")  # 😀 эмодзи

print("\n12. Символ по 8-ричному коду (\\ooo):")
print("\110\145\154\154\157")  # «Hello»

print("\n13. Символ Null (\\0):")
print("До нуля\0после нуля (может не отображаться)")
```

```python
# Шаг 2_4_5
# Тема урока: Строки как списки

# Любую строку в Python можно считать списком отдельных символов, из которых она состоит.

# Примеры работы со строками как с последовательностями (списками):

# Индексация: получение символа по индексу
s = 'Hello, World!'
print(s[0])        # 'H' — первый символ

# Срезы: получение подстроки
print(s[1:-1])     # 'ello, World' — с 1-го до предпоследнего символа
print(s[::3])      # 'Hl r!' — каждый третий символ

# Метод join() — "разредить" строку, вставив разделитель между символами
print(' '.join(s)) # 'H e l l o ,   W o r l d !' — символы строки разделены пробелом

'''
Примечание: В Python строки неизменяемы, но поддерживают те же операции, что и списки: индексацию, срезы, методы len(), in, join() и др.
'''
```
