```python
# Шаг 3_2_6

# Задание 6: Сумма чисел

# Считайте строку (используйте input()), содержащую одно натуральное число n.

# Найдите сумму всех целых чисел от 0 до n включительно, которые:
#   • делятся на 5 без остатка,
#   • но при этом делятся на 3 с остатком.

# Примеры:
#   При n = 100 → сумма = 735
#   При n = 422 → сумма = 11760

# Примечание:
#   Условие "делятся на 5 без остатка, но на 3 с остатком" означает:
#      (x % 5 == 0) and (x % 3 != 0)

n = int(input())

res = 0
for i in range(1, n+1, 1):
    if (i % 5 == 0) and (i % 3 != 0):
        res += i
print(res)
```

```python
# Шаг 3_2_8

# Задание 8: front_x (с пустыми строками)

# Напишите функцию `front_x(words)`, которая на вход принимает список строк и возвращает отсортированный по правилам:

# 1. Слова, начинающиеся с символа "x", идут первыми.
# 2. Остальные слова — в лексикографическом порядке.

# Важно!
#   • Список может содержать пустые строки — их нельзя удалять, и функция не должна "падать".

# Пример:
#   words = ['mix', 'extra', '', 'x-files', 'xyz', 'xapple', 'apple']
#   → результат: ['x-files', 'xapple', 'xyz', '', 'apple', 'extra', 'mix']

# Примечание:
#   • В данной задаче не требуется осуществлять вывод — проверка производится для разных списков!

words = ['mix', 'extra', '', 'x-files', 'xyz', 'xapple', 'apple']

def front_x(words):
    x1 = sorted([a for a in words if a.startswith('x')])
    x2 = sorted([a for a in words if not a.startswith('x')])
    return x1 + x2

front_x(words)
```

```python
# Шаг 3_2_9

# Задание 9: Числа Фибоначчи

# Напишите функцию `fib(n)`, возвращающую n-е число Фибоначчи.

# Последовательность Фибоначчи:
#   F(0) = 0
#   F(1) = 1
#   F(n) = F(n-1) + F(n-2), при n ≥ 2

# Примеры:
#   fib(1) → 1
#   fib(2) → 1
#   fib(5) → 5
#   fib(9) → 34

# Примечание:
#   • В данной задаче не требуется осуществлять вывод — проверка производится для разных чисел!

def fib(n):
    if n < 3:
        return 1
    else:
        return fib(n-1) + fib(n-2)

fib(7)
```

```python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        perv, curr = 0, 1

        for _ in range(2, n+1):
            perv, curr = curr, perv + curr

    return curr

fib(7)
```

```python
fib = lambda n: fib(n-1) + fib(n-2) if n >= 3 else 1

fib(7)
```

```python
def is_prime(n):
    for i in range(2, int(n**0.5)+1):
        if n == 2:
           a = True
        elif n%i != 0:
            i += 1
        else:
            a = False
            break
        a = True
    return a

is_prime(62)
```
