```python
# Шаг 3_2_2

# Задание 2: Производная

# Проверьте (для нескольких разных точек), что производная от e^x совпадает с исходной функцией, т.е. (e^x)' = e^x.

# Для этого реализуйте функцию def_e(x), находящую численное значение производной (с точностью до 3 знаков) в точке x.

# Примечание:
#   • В данной задаче не требуется осуществлять вывод — проверка производится для разных точек!
#   • Пример вывода дан для x = 3 (для понимания, как выглядит результат).

# Подсказка:
#   Численная производная может быть вычислена по формуле:
#      f'(x) ≈ (f(x + h) - f(x)) / h
#   где h — малое число (например, 1e-6).
import math as mt

x0 = 3
# x0 = int(input())

def def_e(x):
    dx = 0.000001
    return round(((mt.exp(x + dx) - mt.exp(x)) / dx), 3)

def_e(x0)
```

```python
# Шаг 3_2_3

# Задание 3: Список — чётные индексы

# Напишите функцию `even_indices(L)`, которая будет возвращать только элементы с чётными индексами.

# Пример:
#   Для L = [1, 1, 2, 3, 5, 8, 13, 21, 34] → результат: [1, 2, 5, 13, 34]

# Примечание:
#   • В данной задаче не требуется осуществлять вывод — проверка производится для разных списков!

L = [1, 1, 2, 3, 5, 8, 13, 21, 34]

def even_indeces(L):
    L2 = []
    c = 0
    for i in L:
        if c % 2 == 0:
            L2.append(i)
        c += 1
        # else:
           # continue
    return L2

even_indeces(L)
```

```python
# Шаг 3_2_4
# Задание 4: Список — чётные элементы

# Напишите функцию `even_elements(L)`, которая возвращает только чётные элементы списка.

# Пример:
#   Для L = [1, 1, 2, 3, 5, 8, 13, 21, 34] → результат: [2, 8, 34]

# Примечание:
#   • В данной задаче не требуется осуществлять вывод — проверка производится для разных списков!

def even_elements(L):
    L2 = []
    c = 0
    for i in L:
        if i % 2 == 0:
            L2.append(i)
        c += 1
        # else:
           # continue
    return L2

even_elements(L)
```
