```python
# Шаг 3_2_2

# Задание 2: Производная

# Проверьте (для нескольких разных точек), что производная от e^x совпадает с исходной функцией, т.е. (e^x)' = e^x.

# Для этого реализуйте функцию def_e(x), находящую численное значение производной (с точностью до 3 знаков) в точке x.

# Примечание:
#   • В данной задаче не требуется осуществлять вывод — проверка производится для разных точек!
#   • Пример вывода дан для x = 3 (для понимания, как выглядит результат).

# Подсказка:
#   Численная производная может быть вычислена по формуле:
#      f'(x) ≈ (f(x + h) - f(x)) / h
#   где h — малое число (например, 1e-6).
import math as mt

x0 = 3
# x0 = int(input())

def def_e(x):
    dx = 0.000001
    return round(((mt.exp(x + dx) - mt.exp(x)) / dx), 3)

def_e(x0)
```
