```python
# Шаг 3_1_7
# Поиск строк с одинаковыми первым и последним символами

words = ['a', 'aa', 'aaa', 'anaconda', 'import', 'cat']
```

```python
def math_ends(L):
    counter = 0
    for word in L:
        if (len(word) > 1) and (word[0] == word[-1]):
            counter += 1
    return counter
```

```python
math_ends(words)
```

```python
# Шаг 3_1_8
# Сортировка списка строк, сперва строки на "x"

words = ['mix', 'extra', 'x-files', 'xyz', 'xapple', 'apple']
```

```python
x_first = []
for word in words:
    if word[0] == "x":
        x_first.append(word)

x_first
```

```python
x_first = [word for word in words if word[0] == "x"]
x_first
```

```python
def front_x(words):
    x_first = [word for word in words if word[0] == "x"]
    not_x_first = [word for word in words if word[0] != "x"]
    return sorted(x_first) + sorted(not_x_first)

front_x(words)

```
