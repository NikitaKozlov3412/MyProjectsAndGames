```python
# –®–∞–≥ 2_7_1

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –§—É–Ω–∫—Ü–∏–∏

# –§—É–Ω–∫—Ü–∏–∏ ‚Äî —ç—Ç–æ –±–ª–æ–∫–∏ –∫–æ–¥–∞, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ –≤—ã–∑—ã–≤–∞—Ç—å –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ.
# –ò—Ö –∏—Å–ø–æ–ª—å–∑—É—é—Ç, –∫–æ–≥–¥–∞ –∫–æ–¥ –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è –∏–ª–∏ –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –≤—ã–¥–µ–ª–∏—Ç—å –ª–æ–≥–∏—á–µ—Å–∫–∏ –∑–∞–≤–µ—Ä—à—ë–Ω–Ω—ã–π –±–ª–æ–∫ –¥–ª—è –ª—É—á—à–µ–≥–æ –ø–æ–Ω–∏–º–∞–Ω–∏—è.

# --- –°—Ç—Ä—É–∫—Ç—É—Ä–∞ —Ñ—É–Ω–∫—Ü–∏–∏ ---

# 1. –û–±—ä—è–≤–ª–µ–Ω–∏–µ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å –∫–ª—é—á–µ–≤–æ–≥–æ —Å–ª–æ–≤–∞ `def`.
# 2. –ü–æ—Å–ª–µ `def` –∏–¥—ë—Ç –∏–º—è —Ñ—É–Ω–∫—Ü–∏–∏ (—Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –¥–∞–≤–∞—Ç—å –æ—Å–º—ã—Å–ª–µ–Ω–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è).
# 3. –í –∫—Ä—É–≥–ª—ã—Ö —Å–∫–æ–±–∫–∞—Ö —É–∫–∞–∑—ã–≤–∞—é—Ç—Å—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ã (–µ—Å–ª–∏ –µ—Å—Ç—å; –µ—Å–ª–∏ –Ω–µ—Ç ‚Äî –ø—É—Å—Ç—ã–µ —Å–∫–æ–±–∫–∏).
# 4. –¢–µ–ª–æ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–∏—à–µ—Ç—Å—è —Å –æ—Ç—Å—Ç—É–ø–æ–º (–∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ —Ü–∏–∫–ª–∞–º –∏ —É—Å–ª–æ–≤–∏—è–º).
# 5. –î–ª—è –≤–æ–∑–≤—Ä–∞—Ç–∞ –∑–Ω–∞—á–µ–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫–ª—é—á–µ–≤–æ–µ —Å–ª–æ–≤–æ `return`.

# –ü—Ä–∏–º–µ—Ä –ø—Ä–æ—Å—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –±–µ–∑ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:
def Hello():
    print('Hello, World!')

# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏:
Hello()  # ‚Üí Hello, World!

# --- –§—É–Ω–∫—Ü–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º ---
def Hello2(name):
    print('Hello,', name, '!', sep='')

Hello2('Aleks')     # ‚Üí Hello, Aleks!
Hello2('UserName')  # ‚Üí Hello, UserName!

# --- –í–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è ---
# –ï—Å–ª–∏ –Ω—É–∂–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–±–æ—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏ ‚Äî –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `return`.

def pow2(x):
    return x**2

y = pow2(5)
print(y)  # ‚Üí 25

# --- –§—É–Ω–∫—Ü–∏—è —Å —É—Å–ª–æ–≤–∏–µ–º –∏ —Å–æ–æ–±—â–µ–Ω–∏–µ–º –æ–± –æ—à–∏–±–∫–µ ---
def sqrt(x):
    if x >= 0:
        return x**0.5
    print('–û–®–ò–ë–ö–ê! –ù–µ —É–¥–∞—ë—Ç—Å—è –∏–∑–≤–ª–µ—á—å –∫–æ—Ä–µ–Ω—å.')

a1 = sqrt(4)   # ‚Üí 2.0
a2 = sqrt(0)   # ‚Üí 0.0
a3 = sqrt(-4)  # ‚Üí None (–Ω–∏—á–µ–≥–æ –Ω–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è, –ø–µ—á–∞—Ç–∞–µ—Ç—Å—è –æ—à–∏–±–∫–∞)

print(a1)  # ‚Üí 2.0
print(a2)  # ‚Üí 0.0
print(a3)  # ‚Üí None

# ‚ö†Ô∏è –í–∞–∂–Ω–æ: –ø–æ—Ä—è–¥–æ–∫ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è!
# –°–Ω–∞—á–∞–ª–∞ –≤—ã–ø–æ–ª–Ω—è—é—Ç—Å—è –≤—Å–µ –≤—ã–∑–æ–≤—ã —Ñ—É–Ω–∫—Ü–∏–π, –ø–æ—Ç–æ–º ‚Äî –≤—ã–≤–æ–¥ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö.
# –ï—Å–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è –Ω–∏—á–µ–≥–æ –Ω–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç (`return` –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç), –æ–Ω–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç `None`.
```

```python
# –®–∞–≥ 2_7_5

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –ó–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é

# –í Python –º–æ–∂–Ω–æ –∑–∞–¥–∞–≤–∞—Ç—å –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é.
# –ï—Å–ª–∏ –ø—Ä–∏ –≤—ã–∑–æ–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è —Ç–∞–∫–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –Ω–µ –ø–µ—Ä–µ–¥–∞–Ω–æ ‚Äî –±—É–¥–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é.

# –ü—Ä–∏–º–µ—Ä:
def sqrt_new(x, n=2):
    return x**(1/n)

# –ü—Ä–∏ –≤—ã–∑–æ–≤–µ:
print(sqrt_new(4))     # ‚Üí 2.0 (n = 2 –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)
print(sqrt_new(27, 3)) # ‚Üí 3.0 (n = 3 –ø–µ—Ä–µ–¥–∞–Ω–æ —è–≤–Ω–æ)

# ‚ö†Ô∏è –í–∞–∂–Ω–æ: –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –±–µ–∑ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –¥–æ–ª–∂–Ω—ã –∏–¥—Ç–∏ **–ø–µ—Ä–≤—ã–µ**!
# –¢–æ –µ—Å—Ç—å –Ω–µ–ª—å–∑—è –ø–∏—Å–∞—Ç—å: def func(a=1, b) ‚Äî —ç—Ç–æ –æ—à–∏–±–∫–∞.

# --- –ò–º–µ–Ω–æ–≤–∞–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã ---

# –ï—Å–ª–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–Ω–æ–≥–æ, –º–æ–∂–Ω–æ –æ–±—Ä–∞—â–∞—Ç—å—Å—è –∫ –Ω–∏–º –ø–æ –∏–º–µ–Ω–∏ –ø—Ä–∏ –≤—ã–∑–æ–≤–µ.
# –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞—Ä—É—à–∞—Ç—å –ø–æ—Ä—è–¥–æ–∫ –ø–µ—Ä–µ–¥–∞—á–∏ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤.

# –ü—Ä–∏–º–µ—Ä:
def func(a, b, c=2):
    return a + b + c

func(b=1, a=5)  # ‚Üí 8 (c = 2 –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)

# –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≤ –±–∏–±–ª–∏–æ—Ç–µ–∫–∞—Ö —Å –±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.
```

```python
# –®–∞–≥ 2_7_7

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –†–µ–∫—É—Ä—Å–∏—è

# –†–µ–∫—É—Ä—Å–∏—è ‚Äî —ç—Ç–æ –∫–æ–≥–¥–∞ —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–∑—ã–≤–∞–µ—Ç —Å–∞–º–∞ —Å–µ–±—è –≤–Ω—É—Ç—Ä–∏ —Å–≤–æ–µ–≥–æ —Ç–µ–ª–∞.

# –ü—Ä–∏–º–µ—Ä: –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ —á–∏—Å–ª–∞ n.
# –ü–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é: n! = 1 √ó 2 √ó ... √ó n

# –ú–æ–∂–Ω–æ –≤—ã—Ä–∞–∑–∏—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ:
#   n! = n √ó (n-1)!

# –î–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã —Ä–µ–∫—É—Ä—Å–∏–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∑–∞–¥–∞—Ç—å **—É—Å–ª–æ–≤–∏–µ –≤—ã—Ö–æ–¥–∞** ‚Äî –∏–Ω–∞—á–µ —Ñ—É–Ω–∫—Ü–∏—è –±—É–¥–µ—Ç –≤—ã–∑—ã–≤–∞—Ç—å —Å–µ–±—è –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ.

# –î–ª—è —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ —É—Å–ª–æ–≤–∏–µ –≤—ã—Ö–æ–¥–∞: 0! = 1 (–ø–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é).

# –ü—Ä–∏–º–µ—Ä —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏:

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# ‚ö†Ô∏è –í–∞–∂–Ω–æ:
# –†–µ–∫—É—Ä—Å–∏—è –Ω–µ –≤—Å–µ–≥–¥–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∏ –ø–∞–º—è—Ç–∏.
# –í Python –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–æ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –Ω–∞ –≥–ª—É–±–∏–Ω—É —Ä–µ–∫—É—Ä—Å–∏–∏ ‚Äî –Ω–µ –±–æ–ª–µ–µ 1000 –≤—ã–∑–æ–≤–æ–≤.
```

```python
# –®–∞–≥ 2_7_8

# –ó–∞–¥–∞–Ω–∏–µ: –†–µ–∞–ª–∏–∑–∞—Ü–∏—è –¥–≤–æ–π–Ω–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞

# –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `dfactorial(n)`, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –¥–≤–æ–π–Ω–æ–π —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞ n.

# –î–≤–æ–π–Ω–æ–π —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª n!! –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤—Å–µ—Ö –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –≤ –æ—Ç—Ä–µ–∑–∫–µ [1, n],
# –∏–º–µ—é—â–∏—Ö —Ç—É –∂–µ —á—ë—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –∏ n:

#   ‚Ä¢ –î–ª—è —á—ë—Ç–Ω–æ–≥–æ n:  n!! = 2 √ó 4 √ó 6 √ó ... √ó n
#   ‚Ä¢ –î–ª—è –Ω–µ—á—ë—Ç–Ω–æ–≥–æ n: n!! = 1 √ó 3 √ó 5 √ó ... √ó n

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –í —ç—Ç–æ–π –∑–∞–¥–∞—á–µ –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏—á–µ–≥–æ —Å—á–∏—Ç—ã–≤–∞—Ç—å –∏ –Ω–∏—á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∞ –ø–µ—á–∞—Ç—å.
# –¢—Ä–µ–±—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∏–º–µ–Ω–µ–º –∏ –ª–æ–≥–∏–∫–æ–π.

def dfactorial(n):
    if n == 0:
        return 1
    elif n == 1:
        return n * dfactorial(n - 1)
    elif n > 1:
        return n * dfactorial(n - 2)
    else:
        return("–ù–µ–ø–æ–¥—Ö–æ–¥—è—â–µ–µ —á–∏—Å–ª–æ")


print(dfactorial(5))
```

```python
# –†–µ—à–µ–Ω–∏–µ –∏–∑ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤

def dfactorial(n):
    pr = 1
    for it in range(n, 0, -2):
        pr *= it
    return pr
```

```python
# –®–∞–≥ 2_7_9

# –ó–∞–¥–∞–Ω–∏–µ: –†–µ–∞–ª–∏–∑–∞—Ü–∏—è –∫—Ä–∞—Ç–Ω–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞

# –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `Kfactorial(n, k)`, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫—Ä–∞—Ç–Ω—ã–π —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞ n —Å –∫—Ä–∞—Ç–Ω–æ—Å—Ç—å—é k.

# –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å 2 –ø–∞—Ä–∞–º–µ—Ç—Ä–∞:
#   1. n ‚Äî —á–∏—Å–ª–æ, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª (–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π).
#   2. k ‚Äî –∫—Ä–∞—Ç–Ω–æ—Å—Ç—å —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ (–Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π, –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é ‚Äî 1).

# –ö—Ä–∞—Ç–Ω—ã–π —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª n!!...! (k –≤–æ—Å–∫–ª–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –∑–Ω–∞–∫–æ–≤) ‚Äî —ç—Ç–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —á–∏—Å–µ–ª:
#   (n - 0) √ó (n - k) √ó (n - 2k) √ó ... –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ —Å–ª–µ–¥—É—é—â–∏–π –º–Ω–æ–∂–∏—Ç–µ–ª—å –Ω–µ —Å—Ç–∞–Ω–µ—Ç –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–µ–Ω 0.

# –£—Å–ª–æ–≤–∏–µ –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Ä–µ–∫—É—Ä—Å–∏–∏: –µ—Å–ª–∏ n <= k, —Ç–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–≤–µ–Ω n.
# –®–∞–≥ —Ä–µ–∫—É—Ä—Å–∏–∏: Kfactorial(n, k) = n * Kfactorial(n - k, k), –µ—Å–ª–∏ n > k.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –í —ç—Ç–æ–π –∑–∞–¥–∞—á–µ –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏—á–µ–≥–æ —Å—á–∏—Ç—ã–≤–∞—Ç—å –∏ –Ω–∏—á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∞ –ø–µ—á–∞—Ç—å.
# –¢—Ä–µ–±—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∏–º–µ–Ω–µ–º –∏ –ª–æ–≥–∏–∫–æ–π.

def Kfactorial(n, k=1):
    pr = 1
    for it in range(n, 0, -k):
        pr *= it
    return pr

print(Kfactorial(7, 2))
```

```python
# –†–µ—à–µ–Ω–∏–µ –∏–∑ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤

def Kfactorial(n, k=1):
    if n < 2:
           return 1

    return n*Kfactorial(n-k,k)

print(Kfactorial(8, 3))
```

```python
# –®–∞–≥ 2_7_10

# –ó–∞–¥–∞–Ω–∏–µ: –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞

# –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `convert(L)`, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –Ω–∞ –≤—Ö–æ–¥ —Å–ø–∏—Å–æ–∫, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ —á–∏—Å–µ–ª –∏ —Å—Ç—Ä–æ–∫ –≤–∏–¥–∞:

# [1, 2, '3', '4', '5', 6]

# –∏ –≤–æ–∑–≤—Ä–∞—â–∞—é—â—É—é —Å–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª (–≤ —Ç–æ–º –∂–µ –ø–æ—Ä—è–¥–∫–µ):

# [1, 2, 3, 4, 5, 6]

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –í —ç—Ç–æ–π –∑–∞–¥–∞—á–µ –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏—á–µ–≥–æ —Å—á–∏—Ç—ã–≤–∞—Ç—å –∏ –Ω–∏—á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∞ –ø–µ—á–∞—Ç—å.
# –¢—Ä–µ–±—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∏–º–µ–Ω–µ–º –∏ –ª–æ–≥–∏–∫–æ–π.

def convert(L):
    L2 = []
    for a in L:
        if type(a) == "str":
            b = int(a[1:-1])
            L2.append(b)
        else:
            b = int(a)
            L2.append(b)
    return L2


convert([1, 2, '3', '4', '5', 6])
```

```python
# –†–µ—à–µ–Ω–∏–µ –∏–∑ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤

def convert(L):
    return [int(i) for i in L]
```

```python
# –®–∞–≥ 2_7_11

# –ó–∞–¥–∞–Ω–∏–µ: –ü–µ—Ä–µ–≤–æ–¥ —á–∏—Å–ª–∞ –∏–∑ –¥–µ—Å—è—Ç–∏—á–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã –≤ —Å–∏—Å—Ç–µ–º—É —Å –æ—Å–Ω–æ–≤–∞–Ω–∏–µ–º p

# –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `translate(n, p=2)`, –∫–æ—Ç–æ—Ä–∞—è:
#   ‚Ä¢ –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä `n` ‚Äî —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ.
#   ‚Ä¢ –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä `p` ‚Äî –æ—Å–Ω–æ–≤–∞–Ω–∏–µ –Ω–æ–≤–æ–π —Å–∏—Å—Ç–µ–º—ã —Å—á–∏—Å–ª–µ–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 2).
#   ‚Ä¢ –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É ‚Äî –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ `n` –≤ —Å–∏—Å—Ç–µ–º–µ —Å—á–∏—Å–ª–µ–Ω–∏—è —Å –æ—Å–Ω–æ–≤–∞–Ω–∏–µ–º `p`.

# –ê–ª–≥–æ—Ä–∏—Ç–º:
#   1. –î–µ–ª–∏—Ç–µ —á–∏—Å–ª–æ `n` –Ω–∞ `p` –∏ –∑–∞–ø–∏—Å—ã–≤–∞–π—Ç–µ –æ—Å—Ç–∞—Ç–æ–∫.
#   2. –ü–æ–≤—Ç–æ—Ä—è–π—Ç–µ —à–∞–≥ 1 —Å —á–∞—Å—Ç–Ω—ã–º –æ—Ç –¥–µ–ª–µ–Ω–∏—è, –ø–æ–∫–∞ –æ–Ω–æ –Ω–µ —Å—Ç–∞–Ω–µ—Ç —Ä–∞–≤–Ω—ã–º 0.
#   3. –°–æ–±–µ—Ä–∏—Ç–µ –≤—Å–µ –æ—Å—Ç–∞—Ç–∫–∏ –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ ‚Äî —ç—Ç–æ –∏ –µ—Å—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

# –ü—Ä–∏–º–µ—Ä—ã:
#   translate(19)     ‚Üí "10011"  (19 –≤ –¥–≤–æ–∏—á–Ω–æ–π)
#   translate(19, 7)  ‚Üí "25"     (19 –≤ —Å–µ–º–µ—Ä–∏—á–Ω–æ–π)
#   translate(10, 2)  ‚Üí "1010"   (10 –≤ –¥–≤–æ–∏—á–Ω–æ–π)

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:
#   ‚Ä¢ –í —ç—Ç–æ–π –∑–∞–¥–∞—á–µ –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏—á–µ–≥–æ —Å—á–∏—Ç—ã–≤–∞—Ç—å –∏ –Ω–∏—á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∞ –ø–µ—á–∞—Ç—å.
#   ‚Ä¢ –¢—Ä–µ–±—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∏–º–µ–Ω–µ–º –∏ –ª–æ–≥–∏–∫–æ–π.

def translate(a, n=2):
    L = []
    L.append(a%n)
    while a//n: #–∏–ª–∏ a//n > 0:
        a = a//n
        b = a%n
        L.insert(0, b)
    return ''.join(map(str, L))




translate(57, 7)
```

```python
# –®–∞–≥ 2_7_12

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π –≤–Ω—É—Ç—Ä–∏ –¥—Ä—É–≥–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π

# –í Python –Ω–µ—Ç –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π –Ω–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π –≤–Ω—É—Ç—Ä–∏ –¥—Ä—É–≥–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π.
# –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–æ–∑–¥–∞–≤–∞—Ç—å —Å–ª–æ–∂–Ω—ã–µ, –Ω–æ —á–∏—Ç–∞–µ–º—ã–µ –∏ –º–æ–¥—É–ª—å–Ω—ã–µ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏.

# –ü—Ä–∏–º–µ—Ä:
def f(x):
    return x**3

def g(x):
    return abs(f(x))  # –≤—ã–∑–æ–≤ f(x) –≤–Ω—É—Ç—Ä–∏ g(x)

def k(x):
    return 2*x + f(x) + g(x)  # –≤—ã–∑–æ–≤—ã f(x) –∏ g(x) –≤–Ω—É—Ç—Ä–∏ k(x)

# –ü—Ä–∏ –≤—ã–∑–æ–≤–µ:
print('f(-1):', f(-1))   # ‚Üí -1
print('g(-1):', g(-1))   # ‚Üí 1 (abs(-1))
print('k(-1):', k(-1))   # ‚Üí -2 (2*(-1) + (-1) + 1)

# üí° –ó–∞–º–µ—Ç–∏–º:
#   ‚Ä¢ `f(x)` –∏ `g(x)` –∏–º–µ—é—Ç –ø—Ä–æ—Ç–∏–≤–æ–ø–æ–ª–æ–∂–Ω—ã–µ –∑–Ω–∞–∫–∏ –ø—Ä–∏ `x < 0`, –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ ‚Äî –ø—Ä–∏ `x > 0`.
#   ‚Ä¢ –§—É–Ω–∫—Ü–∏—è `k(x)` —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ø—Ä–æ—â–µ (`2x`) –ø—Ä–∏ `x < 0`, –∏ —Å–ª–æ–∂–Ω–µ–µ (`2x + 2x¬≥`) –ø—Ä–∏ `x > 0`.

# –¢–∞–∫–∞—è –≤–ª–æ–∂–µ–Ω–Ω–æ—Å—Ç—å —Ñ—É–Ω–∫—Ü–∏–π ‚Äî –Ω–æ—Ä–º–∞–ª—å–Ω–∞—è –ø—Ä–∞–∫—Ç–∏–∫–∞.
# –û–Ω–∞ –ø–æ–º–æ–≥–∞–µ—Ç —Ä–∞–∑–±–∏–≤–∞—Ç—å –∑–∞–¥–∞—á—É –Ω–∞ –ª–æ–≥–∏—á–µ—Å–∫–∏–µ —á–∞—Å—Ç–∏ –∏ –¥–µ–ª–∞–µ—Ç –∫–æ–¥ –±–æ–ª–µ–µ –ø–æ–Ω—è—Ç–Ω—ã–º.
```

```python
# –®–∞–≥ 2_7_13

# –ó–∞–¥–∞–Ω–∏–µ: –§–∞–∫—Ç–æ—Ä–∏–∞–ª –∏ —Å—É–ø–µ—Ä—Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª

# –ù–∞–ø–∏—à–∏—Ç–µ 2 —Ñ—É–Ω–∫—Ü–∏–∏:

# 1. `factorial(n)` ‚Äî –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞ n:
#    factorial(n) = n! = 1 √ó 2 √ó 3 √ó ... √ó n

# 2. `sf(n)` ‚Äî –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—É–ø–µ—Ä—Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞ n:
#    sf(n) = 1! √ó 2! √ó 3! √ó ... √ó n!

# –í —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ —Ñ—É–Ω–∫—Ü–∏–∏ `sf(n)` –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `factorial`, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –¥—É–±–ª–∏—Ä–æ–≤–∞–Ω–∏—è –∫–æ–¥–∞.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:
#   ‚Ä¢ –í —ç—Ç–æ–π –∑–∞–¥–∞—á–µ –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏—á–µ–≥–æ —Å—á–∏—Ç—ã–≤–∞—Ç—å –∏ –Ω–∏—á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∞ –ø–µ—á–∞—Ç—å.
#   ‚Ä¢ –¢—Ä–µ–±—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –¥–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∏–º–µ–Ω–∞–º–∏ –∏ –ª–æ–≥–∏–∫–æ–π.

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

def sf(n):
    if n == 0:
        return factorial(1)
    else:
        return factorial(n) * sf(n-1)

n = int(input())
print(factorial(n), sf(n))
```

```python
# üåü –ú–∏–Ω–∏-—É—Ä–æ–∫: –ö–æ–º–ø–æ–∑–∏—Ü–∏—è —Ñ—É–Ω–∫—Ü–∏–π –≤ Python

# –ö–æ–º–ø–æ–∑–∏—Ü–∏—è —Ñ—É–Ω–∫—Ü–∏–π ‚Äî —ç—Ç–æ —Å–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –ø—É—Ç—ë–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –¥–≤—É—Ö (–∏–ª–∏ –±–æ–ª–µ–µ) —Å—É—â–µ—Å—Ç–≤—É—é—â–∏—Ö.

# –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏: (f ‚àò g)(x) = f(g(x))

# –í Python —ç—Ç–æ –ª–µ–≥–∫–æ —Ä–µ–∞–ª–∏–∑—É–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü–∏–π –∏–ª–∏ –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö –≤—ã–∑–æ–≤–æ–≤.

# --- –ü—Ä–∏–º–µ—Ä ---

def h(x):
    return x + 1      # "–ø—Ä–∏–±–∞–≤–∏—Ç—å 1"

def g(x):
    return 3 * x      # "—É–º–Ω–æ–∂–∏—Ç—å –Ω–∞ 3"

# –ü—Ä—è–º–æ–π –≤—ã–∑–æ–≤:
result1 = g(h(1))     # ‚Üí g(2) ‚Üí 6

# –°–æ–∑–¥–∞–¥–∏–º –∫–æ–º–ø–æ–∑–∏—Ü–∏—é –æ–¥–∏–Ω —Ä–∞–∑ –∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ:
def composition(f1, f2):
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é f1 ‚àò f2: —Å–Ω–∞—á–∞–ª–∞ f2, –ø–æ—Ç–æ–º f1"""
    return lambda x: f1(f2(x))

gh = composition(g, h)   # gh(x) = g(h(x))
result2 = gh(1)          # ‚Üí 6

print(result1 == result2)  # True

# --- –ó–∞—á–µ–º —ç—Ç–æ –Ω—É–∂–Ω–æ? ---
# ‚úÖ –ò–∑–±–µ–≥–∞–µ–º –¥—É–±–ª–∏—Ä–æ–≤–∞–Ω–∏—è: gh –º–æ–∂–Ω–æ –≤—ã–∑—ã–≤–∞—Ç—å –º–Ω–æ–≥–æ —Ä–∞–∑ –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏—è g(h(...))
# ‚úÖ –ß–∏—Ç–∞–µ–º–æ—Å—Ç—å: gh = g ‚àò h –≤—ã—Ä–∞–∂–∞–µ—Ç –Ω–∞–º–µ—Ä–µ–Ω–∏–µ —è—Å–Ω–µ–µ, —á–µ–º –≤–ª–æ–∂–µ–Ω–Ω—ã–µ –≤—ã–∑–æ–≤—ã
# ‚úÖ –ì–∏–±–∫–æ—Å—Ç—å: –ª–µ–≥–∫–æ –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞—Ç—å —Ä–∞–∑–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏:
hg = composition(h, g)   # h(g(x)) = 3*x + 1
print(hg(1))  # ‚Üí 4

# üí° –ö–æ–º–ø–æ–∑–∏—Ü–∏—è ‚Äî –º–æ—â–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Å–ª–æ–∂–Ω—ã—Ö –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π
#    –∏–∑ –ø—Ä–æ—Å—Ç—ã—Ö, –ø–µ—Ä–µ–∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö ¬´—Å—Ç—Ä–æ–∏—Ç–µ–ª—å–Ω—ã—Ö –±–ª–æ–∫–æ–≤¬ª.
```

```python
# –®–∞–≥ 2_7_14

# –ó–∞–¥–∞–Ω–∏–µ: –ò–Ω–¥–µ–∫—Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —Ü–µ–ª–æ–≥–æ —á–∏—Å–ª–∞

# –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `maxId(L)`, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –Ω–∞ –≤—Ö–æ–¥ —Å–ø–∏—Å–æ–∫, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ —á–∏—Å–µ–ª –∏ —Å—Ç—Ä–æ–∫ –≤–∏–¥–∞:

# [1, 2, '42', '3', '4', '5', 6, 13]

# –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞:
#   ‚Ä¢ –ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ä–∞–Ω–µ–µ –Ω–∞–ø–∏—Å–∞–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é `convert(L)` (–∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Å—Ç—Ä–æ–∫–∏-—á–∏—Å–ª–∞ –≤ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞),
#   ‚Ä¢ –ù–∞–π—Ç–∏ **–∏–Ω–¥–µ–∫—Å** –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —Ü–µ–ª–æ–≥–æ —á–∏—Å–ª–∞ –≤ —Å–ø–∏—Å–∫–µ (–±–µ–∑ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π),
#   ‚Ä¢ –í–µ—Ä–Ω—É—Ç—å —ç—Ç–æ—Ç –∏–Ω–¥–µ–∫—Å.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:
#   ‚Ä¢ –í —ç—Ç–æ–π –∑–∞–¥–∞—á–µ –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏—á–µ–≥–æ —Å—á–∏—Ç—ã–≤–∞—Ç—å –∏ –Ω–∏—á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∞ –ø–µ—á–∞—Ç—å.
#   ‚Ä¢ –¢—Ä–µ–±—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∏–º–µ–Ω–µ–º –∏ –ª–æ–≥–∏–∫–æ–π.

# –ü—Ä–∏–º–µ—Ä—ã:
#   maxId([1, 2, '42', '3', '4', '5', 6, 13]) ‚Üí 2 (–ø–æ—Ç–æ–º—É —á—Ç–æ 42 ‚Äî —Å–∞–º–æ–µ –±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ, –æ–Ω–æ –Ω–∞ –ø–æ–∑–∏—Ü–∏–∏ 2)
#   maxId(['0', 1, 2, '3', '4', '5', 6, '666', 42]) ‚Üí 7 (–ø–æ—Ç–æ–º—É —á—Ç–æ 666 ‚Äî —Å–∞–º–æ–µ –±–æ–ª—å—à–æ–µ, –æ–Ω–æ –Ω–∞ –ø–æ–∑–∏—Ü–∏–∏ 7)

def convert(L):
    L2 = []
    for a in L:
        if type(a) == "str":
            b = int(a[1:-1])
            L2.append(b)
        else:
            b = int(a)
            L2.append(b)
    return L2


def maxId(L):
    # return L.index(max(convert(L)))
    return convert(L).index(max(convert(L)))

maxId([1, 2, '42', '3', '4', '5', 6, 13])
```
