```python
# –®–∞–≥ 2_6_1

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –¶–∏–∫–ª—ã

# –¶–∏–∫–ª—ã ‚Äî —ç—Ç–æ —É–ø—Ä–∞–≤–ª—è—é—â–∏–µ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏ –≤ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–∏,
# –ø—Ä–µ–¥–Ω–∞–∑–Ω–∞—á–µ–Ω–Ω—ã–µ –¥–ª—è –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ–≥–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –±–ª–æ–∫–∞ –∫–æ–¥–∞.

# –ö–æ–Ω–µ—á–Ω–æ, –º–æ–∂–Ω–æ –ø—Ä–æ—Å—Ç–æ –ø—Ä–æ–¥—É–±–ª–∏—Ä–æ–≤–∞—Ç—å –∫–æ–¥ –Ω—É–∂–Ω–æ–µ —á–∏—Å–ª–æ —Ä–∞–∑...
# –ù–æ —ç—Ç–æ –±—ã—Å—Ç—Ä–æ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –Ω–µ—É–¥–æ–±–Ω—ã–º –∏ –Ω–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–º.

# –¶–∏–∫–ª—ã —Ä–µ—à–∞—é—Ç —Ç—Ä–∏ –∫–ª—é—á–µ–≤—ã–µ –ø—Ä–æ–±–ª–µ–º—ã:

# 1. ‚ùì –ß—Ç–æ –µ—Å–ª–∏ —á–∏—Å–ª–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π –≤–µ–ª–∏–∫–æ?
#    ‚Üí –¢–æ–≥–¥–∞ –¥—É–±–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–æ–¥–∞ —Å–¥–µ–ª–∞–µ—Ç –ø—Ä–æ–≥—Ä–∞–º–º—É –≥—Ä–æ–º–æ–∑–¥–∫–æ–π –∏ –Ω–µ—á–∏—Ç–∞–µ–º–æ–π.

# 2. ‚ùì –ß—Ç–æ –µ—Å–ª–∏ —á–∏—Å–ª–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π –∑–∞—Ä–∞–Ω–µ–µ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ?
#    ‚Üí –ù–∞–ø—Ä–∏–º–µ—Ä, –Ω—É–∂–Ω–æ —á–∏—Ç–∞—Ç—å –≤–≤–æ–¥, –ø–æ–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–µ –≤–≤–µ–¥—ë—Ç "—Å—Ç–æ–ø".

# 3. ‚ùì –ß—Ç–æ –µ—Å–ª–∏ –ø—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —É—Å–ª–æ–≤–∏—è—Ö –Ω—É–∂–Ω–æ –ø—Ä–µ—Ä–≤–∞—Ç—å –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–µ?
#    ‚Üí –¶–∏–∫–ª—ã –ø–æ–∑–≤–æ–ª—è—é—Ç –≥–∏–±–∫–æ —É–ø—Ä–∞–≤–ª—è—Ç—å —Ö–æ–¥–æ–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è (—á–µ—Ä–µ–∑ break, continue –∏ —Ç.–¥.).

# –í Python —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –¥–≤–∞ –æ—Å–Ω–æ–≤–Ω—ã—Ö –≤–∏–¥–∞ —Ü–∏–∫–ª–æ–≤:

# ‚Ä¢ for ‚Äî –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π (—Å—Ç—Ä–æ–∫, —Å–ø–∏—Å–∫–æ–≤, –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –∏ –¥—Ä.)
# ‚Ä¢ while ‚Äî –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –ø–æ–∫–∞ –∑–∞–¥–∞–Ω–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ –æ—Å—Ç–∞—ë—Ç—Å—è –∏—Å—Ç–∏–Ω–Ω—ã–º (True)

# –ü—Ä–∏–º–µ—Ä—ã:

# –¶–∏–∫–ª for:
for i in range(3):
    print("–ü—Ä–∏–≤–µ—Ç")

# –¶–∏–∫–ª while:
count = 0
while count < 3:
    print("–ü–æ–∫–∞")
    count += 1

# –í –¥–∞–ª—å–Ω–µ–π—à–µ–º –º—ã –ø–æ–¥—Ä–æ–±–Ω–æ –∏–∑—É—á–∏–º —Å–∏–Ω—Ç–∞–∫—Å–∏—Å, –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –∫–∞–∂–¥–æ–≥–æ –∏–∑ –Ω–∏—Ö.
```

```python
# –®–∞–≥ 2_6_3

# –ó–∞–¥–∞–Ω–∏–µ: –í—ã–≤–æ–¥ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —á–∏—Å–µ–ª

# –°—á–∏—Ç–∞–π—Ç–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n.

# –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ –ø–µ—á–∞—Ç—å –≤—Å–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –æ—Ç 0 –¥–æ n –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ,
# –∫–∞–∂–¥–æ–µ —á–∏—Å–ª–æ ‚Äî —Å –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏.

'''
> **–ü—Ä–∏–º–µ—Ä**:
> –ü—Ä–∏ –≤–≤–æ–¥–µ `3` –ø—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏:
> 0
> 1
> 2
> 3
'''

n = int(input())
i = 0
while i <= n:
    print(i)
    i = i+1


# —Ä–µ—à–µ–Ω–∏–µ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É (–±–µ–∑ —Ü–∏–∫–ª–∞ while):

# print(*range(int(input())+1),sep='\n')
```

```python
# –®–∞–≥ 2_6_4

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –¶–∏–∫–ª for

# –¶–∏–∫–ª `for` –ø–æ–∑–≤–æ–ª—è–µ—Ç "–ø—Ä–æ–π—Ç–∏" –ø–æ –ª—é–±–æ–º—É –∏—Ç–µ—Ä–∏—Ä—É–µ–º–æ–º—É –æ–±—ä–µ–∫—Ç—É ‚Äî —Ç–æ –µ—Å—Ç—å –ø–æ —Ç–æ–º—É,
# —á—Ç–æ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å –ø–æ—ç–ª–µ–º–µ–Ω—Ç–Ω–æ: —Å–ø–∏—Å–∫–∏, —Å—Ç—Ä–æ–∫–∏, –¥–∏–∞–ø–∞–∑–æ–Ω—ã, –∫–æ—Ä—Ç–µ–∂–∏ –∏ –¥—Ä.

# –ë–∞–∑–æ–≤—ã–π —Å–∏–Ω—Ç–∞–∫—Å–∏—Å:
# for element in iterator:
#     ...

# –ì–¥–µ:
#   - `for` ‚Äî –∫–ª—é—á–µ–≤–æ–µ —Å–ª–æ–≤–æ, –æ–±–æ–∑–Ω–∞—á–∞—é—â–µ–µ –Ω–∞—á–∞–ª–æ —Ü–∏–∫–ª–∞.
#   - `element` ‚Äî –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è, –≤ –∫–æ—Ç–æ—Ä—É—é –Ω–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç.
#   - `in` ‚Äî –∫–ª—é—á–µ–≤–æ–µ —Å–ª–æ–≤–æ, —É–∫–∞–∑—ã–≤–∞—é—â–µ–µ, –ø–æ –∫–∞–∫–æ–º—É –æ–±—ä–µ–∫—Ç—É –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –∏—Ç–µ—Ä–∞—Ü–∏—è.
#   - `iterator` ‚Äî –∏—Ç–µ—Ä–∏—Ä—É–µ–º—ã–π –æ–±—ä–µ–∫—Ç (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å–ø–∏—Å–æ–∫, —Å—Ç—Ä–æ–∫–∞).

# –ü—Ä–∏–º–µ—Ä—ã:

# –ü–µ—Ä–µ–±–æ—Ä —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞:
L = [1, 'two', 'tres', '—á–µ—Ç—ã—Ä–µ']
for element in L:
    print(element)

# –ü–µ—Ä–µ–±–æ—Ä —Å–∏–º–≤–æ–ª–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ:
word = 'Hello!'
for symbol in word:
    print(symbol)

# --- –í–∞–∂–Ω–æ–µ –ø—Ä–∏–º–µ—á–∞–Ω–∏–µ ---

# –ù–µ —Å—Ç–æ–∏—Ç –∏–∑–º–µ–Ω—è—Ç—å –∏—Ç–µ—Ä–∏—Ä—É–µ–º—ã–π –æ–±—ä–µ–∫—Ç (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å–ø–∏—Å–æ–∫) –≤–æ –≤—Ä–µ–º—è –µ–≥–æ –ø–µ—Ä–µ–±–æ—Ä–∞ ‚Äî
# —ç—Ç–æ –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –Ω–µ–ø—Ä–µ–¥—Å–∫–∞–∑—É–µ–º—ã–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∏–ª–∏ –æ—à–∏–±–∫–∞–º.

# –°–æ–≤–µ—Ç: –µ—Å–ª–∏ –Ω—É–∂–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–±–æ—Ç—ã —Ü–∏–∫–ª–∞ ‚Äî —Å–æ—Ö—Ä–∞–Ω—è–π—Ç–µ –µ–≥–æ –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å–ø–∏—Å–æ–∫.
```

```python
# –®–∞–≥ 2_6_7

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –¶–∏–∫–ª for –¥–ª—è —Ñ–∞–π–ª–æ–≤

# –û–¥–Ω–∞ –∏–∑ —Å–∞–º—ã—Ö –ø–æ–ª–µ–∑–Ω—ã—Ö –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏–π ‚Äî —Ü–∏–∫–ª `for` –¥–ª—è –ø–æ—Å—Ç—Ä–æ—á–Ω–æ–≥–æ —á—Ç–µ–Ω–∏—è —Ñ–∞–π–ª–∞.

# –ü–æ—Å–ª–µ –æ—Ç–∫—Ä—ã—Ç–∏—è —Ñ–∞–π–ª–∞ –¥–ª—è —á—Ç–µ–Ω–∏—è –µ–≥–æ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ –º–æ–∂–Ω–æ:
#   - —Å—á–∏—Ç–∞—Ç—å —Å—Ä–∞–∑—É –≤ –æ–¥–Ω—É –±–æ–ª—å—à—É—é —Å—Ç—Ä–æ–∫—É, –∏–ª–∏
#   - —á–∏—Ç–∞—Ç—å –ø–æ—Å—Ç—Ä–æ—á–Ω–æ ‚Äî –∫–∞–∂–¥—ã–π —Ä–∞–∑ –ø–æ–ª—É—á–∞—è —Ç–æ–ª—å–∫–æ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É (–¥–æ —Å–∏–º–≤–æ–ª–∞ `\n`).

# –ü—Ä–∏–º–µ—Ä:
f = open('–î–≤—É—Ö—Ç—ã—Å—è—á–Ω—ã–µ.txt', 'r', encoding='utf-8')  # –æ—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª –Ω–∞ —á—Ç–µ–Ω–∏–µ
# –ø—Ä–æ encoding='utf-8' –≤ —É—Ä–æ–∫–µ –Ω–µ –±—ã–ª–æ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ
for line in f:
    print(line)
f.close()  # –∑–∞–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª

# –ß—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç:
#   ‚Ä¢ –û—Ç–∫—Ä—ã–≤–∞–µ—Ç—Å—è —Ñ–∞–π–ª —Å –∏–º–µ–Ω–µ–º `filename.txt`.
#   ‚Ä¢ –§–∞–π–ª –¥–æ–ª–∂–µ–Ω –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –≤ —Ç–æ–º –∂–µ –∫–∞—Ç–∞–ª–æ–≥–µ, —á—Ç–æ –∏ —Å–∫—Ä–∏–ø—Ç.
#   ‚Ä¢ –ü–∞—Ä–∞–º–µ—Ç—Ä `'r'` –æ–∑–Ω–∞—á–∞–µ—Ç "—á—Ç–µ–Ω–∏–µ" ‚Äî —Ñ–∞–π–ª –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏–∑–º–µ–Ω—ë–Ω.
#   ‚Ä¢ –í —Ü–∏–∫–ª–µ `for` –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è `line` –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É –∏–∑ —Ñ–∞–π–ª–∞.
#   ‚Ä¢ –ü–æ—Å–ª–µ –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è —Ä–∞–±–æ—Ç—ã —Å —Ñ–∞–π–ª–æ–º –µ–≥–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∑–∞–∫—Ä—ã—Ç—å (`close()`).

# ‚ö†Ô∏è –í–∞–∂–Ω–æ:
# –§—É–Ω–∫—Ü–∏—è `print()` –¥–æ–±–∞–≤–ª—è–µ—Ç –ø–µ—Ä–µ–Ω–æ—Å —Å—Ç—Ä–æ–∫–∏ –ø–æ—Å–ª–µ –≤—ã–≤–æ–¥–∞, –∞ –∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –∏–∑ —Ñ–∞–π–ª–∞ —É–∂–µ —Å–æ–¥–µ—Ä–∂–∏—Ç `\n` ‚Äî
# –ø–æ—ç—Ç–æ–º—É –º–µ–∂–¥—É —Å—Ç—Ä–æ–∫–∞–º–∏ –≤–æ–∑–Ω–∏–∫–∞–µ—Ç –¥–≤–æ–π–Ω–æ–π –ø–µ—Ä–µ–Ω–æ—Å.

# –û —Ä–∞–±–æ—Ç–µ —Å —Ñ–∞–π–ª–∞–º–∏ –º—ã –ø–æ–≥–æ–≤–æ—Ä–∏–º –ø–æ–¥—Ä–æ–±–Ω–µ–µ –ø–æ–∑–∂–µ.
```

```python
# –®–∞–≥ 2_6_8

# –¢–µ–º–∞ —É—Ä–æ–∫–∞: –û–ø–µ—Ä–∞—Ç–æ—Ä—ã break –∏ continue

# --- Break ---

# –û–ø–µ—Ä–∞—Ç–æ—Ä `break` –ø–æ–∑–≤–æ–ª—è–µ—Ç **–Ω–µ–º–µ–¥–ª–µ–Ω–Ω–æ –ø—Ä–µ—Ä–≤–∞—Ç—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ü–∏–∫–ª–∞** (–∫–∞–∫ `for`, —Ç–∞–∫ –∏ `while`).
# –ß–∞—â–µ –≤—Å–µ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –≤—ã–π—Ç–∏ –∏–∑ —Ü–∏–∫–ª–∞ –ø—Ä–∏ –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏–∏ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–≥–æ —É—Å–ª–æ–≤–∏—è.

# –ü—Ä–∏–º–µ—Ä 1 (for):
for i in range(10):
    print(i)
    if i > 5:
        break

# –ü—Ä–∏–º–µ—Ä 2 (while):
i = 0
while i < 10:
    print(i)
    i += 1
    if i > 5:
        break

# –í –æ–±–æ–∏—Ö —Å–ª—É—á–∞—è—Ö —Ü–∏–∫–ª –∑–∞–≤–µ—Ä—à–∞–µ—Ç—Å—è, –∫–∞–∫ —Ç–æ–ª—å–∫–æ —É—Å–ª–æ–≤–∏–µ `i > 5` —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –∏—Å—Ç–∏–Ω–Ω—ã–º.

# --- Continue ---

# –û–ø–µ—Ä–∞—Ç–æ—Ä `continue` —Ç–∞–∫–∂–µ —Ä–∞–±–æ—Ç–∞–µ—Ç –≤ `for` –∏ `while`, –Ω–æ **–Ω–µ –≤—ã—Ö–æ–¥–∏—Ç –∏–∑ —Ü–∏–∫–ª–∞**.
# –û–Ω –ø—Ä–æ—Å—Ç–æ **–ø—Ä–µ—Ä—ã–≤–∞–µ—Ç —Ç–µ–∫—É—â—É—é –∏—Ç–µ—Ä–∞—Ü–∏—é** –∏ –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç –∫ —Å–ª–µ–¥—É—é—â–µ–π.

# –ü—Ä–∏–º–µ—Ä:
for i in 'hello world':
    if i == 'o':
        continue
    print(i * 2, end='')

# –†–µ–∑—É–ª—å—Ç–∞—Ç: "hheellll   wwrlldd" ‚Äî —Å–∏–º–≤–æ–ª 'o' –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç—Å—è, –æ—Å—Ç–∞–ª—å–Ω—ã–µ —É–¥–≤–∞–∏–≤–∞—é—Ç—Å—è.

```

```python
# –®–∞–≥ 2_6_9

# –ó–∞–¥–∞–Ω–∏–µ: –°—É–º–º–∞ —á–∏—Å–µ–ª –¥–æ "The End"

# –°—á–∏—Ç—ã–≤–∞–π—Ç–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –∏–∑ –ø–æ—Ç–æ–∫–∞ –≤–≤–æ–¥–∞ –ø–æ –æ–¥–Ω–æ–º—É, –ø–æ–∫–∞ –Ω–µ –≤—Å—Ç—Ä–µ—Ç–∏—Ç–µ —Å—Ç—Ä–æ–∫—É "The End".

# –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ –ø–µ—á–∞—Ç—å —Å—É–º–º—É –≤—Å–µ—Ö —Å—á–∏—Ç–∞–Ω–Ω—ã—Ö —á–∏—Å–µ–ª.

sum = 0
a = 1
while a:
    a = input()
    if a == "The End":
        break
    b = int(a)
    sum += b
print(sum)
```

```python
# –î—Ä—É–≥–æ–µ —Ä–µ—à–µ–Ω–∏–µ (–∏–∑ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤):
s = input()
n = 0
while s != 'The End':
    n += int(s)
    s = input()
print(n)
```
