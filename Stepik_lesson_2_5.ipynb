```python
# Шаг 2_5_2

# Тема урока: Логические операторы

# Логические операторы сравнивают выражения слева и справа от оператора
# и возвращают один из двух результатов:
#   - True (истина) — если выражение правдиво
#   - False (ложь) — если выражение ошибочно

# --- "Математические" операторы сравнения ---
# Сравнивают значения (не только числа, но и другие типы)
# ==  "равно"
# !=  "не равно"
# >   "больше"
# >=  "больше или равно"
# <   "меньше"
# <=  "меньше или равно"

# Примеры:
print(5 == 5)      # True
print(3 != 4)      # True
print(7 > 2)       # True
print(10 <= 10)    # True

# --- Логические операторы ---
# Позволяют комбинировать условия

# and — логическое "и": оба выражения должны быть верны
x = 7
result_and = x > 5 and x % 2 != 0  # True and True → True
print(result_and)  # True

# or — логическое "или": хотя бы одно выражение должно быть верно
result_or = x < 5 or x % 2 == 0    # False or False → False
print(result_or)   # False

# not — инвертирование: меняет True на False и наоборот
result_not = not (x > 10)          # not False → True
print(result_not)  # True

# --- Скобки для регулирования порядка операций ---
# Аналогично математике: скобки задают приоритет
complex_condition = (x > 5 and x < 10) or (x == 15)
print(complex_condition)  # True (т.к. x=7 подходит под первое условие)

# --- Оператор in ---
# Проверяет, содержится ли элемент в последовательности (строке, списке и т.д.)

# В строке:
print('a' in 'abracadabra')  # True — буква 'a' есть в строке

# В списке:
print(2 in [1, 3, 5])        # False — числа 2 нет в списке

# Экспериментируйте с разными комбинациями!
```

```python
# Шаг 2_5_3

# Тема урока: Римские цифры

# Считайте число (см. таблицу) в арабской записи и выведите на печать,
# соответствующую ему римскую цифру:

# Соответствие арабских чисел римским:
# 1  → I
# 5  → V
# 10 → X
# 50 → L
# 100 → C
# 500 → D
# 1000 → M

# Примеры:
# Sample Input 1: 1   → Sample Output 1: I
# Sample Input 2: 5   → Sample Output 2: V

# Задание: написать программу, которая по введённому арабскому числу
# из предложенного списка выводит соответствующую римскую цифру.

num = int(input())
if num == 1:
    print('I')
elif num == 5:
    print('V')
elif num == 10:
    print('X')
elif num == 50:
    print('L')
elif num == 100:
    print('C')
elif num == 500:
    print('D')
elif num == 1000:
    print('M')
else:
    print('Нет соответствия с римскими цифрами')
```

```python
# Шаг 2_5_5

# Тема урока: Многоуровневые условия

# Внутри блока кода, выполняющегося при одном из условий,
# всегда можно написать столько новых условий, сколько необходимо.

# Пример вложенных условий:
if test1:
    state1
    if test2:
        state2
        if test3:
            state3

# Каждый новый уровень вложенности добавляет отступ — это называется "лесенка".
# Но не стоит слишком увлекаться такими конструкциями!

# ❗ Почему?
# Больше 2–3 уровней вложенности становится сложно читать и анализировать человеку.
# Код теряет читаемость, возрастает вероятность ошибок.

# ✅ Что делать?
# Если возникает сложная многоуровневая конструкция — возможно, лучше оформить её в виде функции.
# Функция скрывает детали, делает код модульным и понятным.

# Пример рефакторинга:
# Вместо:
# if a:
#     if b:
#         if c:
#             ...

# Лучше:
# def check_conditions(a, b, c):
#     if a and b and c:
#         return True
#     return False

# Использование:
# if check_conditions(a, b, c):
#     ...
```

```python
# Шаг 2_5_6

# Тема урока: Обработка деления на ноль и округление

# Считайте строку.
# Если она содержит число 0, выведите на печать фразу "Division by zero!"

# Если она содержит число, отличное от 0,
# то считайте ещё одну строку, содержащую число,
# и напечатайте результат деления 2-го числа на 1-е.
# Результат деления округлите до 1 знака после запятой (используйте round).

# Примеры:

# Sample Input 1:
# 0
# Sample Output 1:
# Division by zero!

# Sample Input 2:
# 3
# 4
# Sample Output 2:
# 1.3

a = int(input())

if a == 0:
    print('Division by zero!')
else:
    b = int(input())
    c = round(b/a, 1)
    print(c)
```

```python
# Шаг 2_5_6 (ещё один способ, но без указания условия "if a == 0")
a = int(input())

if not a:
    print('Division by zero!')
else:
    b = int(input())
    c = round(b/a, 1)
    print(c)
```

```python
# Шаг 2_5_7

# Тема урока: Приведение типов в логических выражениях

# Несмотря на динамическую типизацию в Python, при сравнении переменных
# **не происходит неявного приведения типов**, как это делается, например, в PHP.
# Строка и число — разные типы, и они **не будут равны** друг другу.

# Пример:
# 2 == '2' → False (число ≠ строка)

# Однако! Многие типы данных могут **самостоятельно участвовать в логических выражениях**
# (например, в `if`) **без явного использования логических операторов**.
# При этом:
#   - "пустые" или "нулевые" значения соответствуют False
#   - "не пустые" значения соответствуют True

# Примеры:

# Пустая строка → False
x = ''
if x:
    print(True)
else:
    print(False)  # → False

# Непустая строка → True
x = 'dsfds'
if x:
    print(True)   # → True
else:
    print(False)

# Число 0 → False
x = 0
if x:
    print(True)
else:
    print(False)  # → False

# Ненулевое число → True
x = 5
if x:
    print(True)   # → True
else:
    print(False)

# Пустой список → False
x = []
if x:
    print(True)
else:
    print(False)  # → False

# Непустой список → True
x = [[]]  # список с одним элементом (даже если этот элемент — пустой список)
if x:
    print(True)   # → True
else:
    print(False)

# --- Комбинируя приведение типов с логическими операторами ---
# Можно получать сложные, но лаконичные условия:

a = 1
b = 1
if a and b:  # оба числа "истинны" (не нулевые)
    print("Оба числа не нулевые")

# --- Про Python way ---

# В Python **не принято** писать:
# if some == 0:     # ❌ избыточно
#     ...

# или

# if string != '':  # ❌ лучше использовать логическое значение строки
#     ...

# Вместо этого пишут:
# if not some:      # ✅ "если some — пустое/нулевое"
# if string:        # ✅ "если string — не пустая"

# Это более питоновский (Pythonic) стиль — короткий, читаемый и безопасный.
```
