```python
grouped_df =  df.groupby(['adm_district', 'city_district']) \
    .agg({'lattitude': 'mean', 'longitude': 'mean'}) \
    .reset_index()
```

```python
pivot_df = grouped_df.pivot(index='adm_district', columns= 'city_district', values = 'lattitude').fillna(0)
```
