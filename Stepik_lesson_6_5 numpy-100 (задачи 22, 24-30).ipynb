```python
# Шаг 6_5_2

# 22 Нормализация

# Используйте формулу x_new = (x - μ) / σ,
# где:
#   - x — старое значение
#   - x_new — новое значение
#   - μ — среднее значение (по всей матрице)
#   - σ — стандартное отклонение (по всей матрице)

# для того чтобы "нормализовать" матрицу.

# Мы пока не будем углубляться в математическую суть этой операции, важно лишь, что оба значения для матрицы могут быть вычислены стандартными функциями питу и вам не нужно звать на помощь статистику)

# Пример:
#   [[1, 2],
#    [2, 1]]
#   μ = (1+2+2+1)/4 = 1.5
#   σ = √(1/4 * Σ(xi - μ)²) = 0.5

# Переменная Y содержит исходную матрицу.

# В переменную Z сохраните её "нормализованный" вид, округлив каждое значение в матрице до 2 знака после запятой с помощью numpy.around.

# Примечание. В этой задаче не нужно ничего выводить на печать. Только создать матрицу Z.

import numpy as np

Y = np.array([[99,11,55],[33,66,99]])
me = np.mean(Y)
std = np.std(Y)
print(me, std)

Z = np.round((Y-me) / std, 2)
print(Z)
```

```python
# Шаг 6_5_4

# 24.2 Умножение матриц

# Даны 2 матрицы: A и B.

# Найдите произведение матриц и сохраните его в переменную Z.

# Если найти произведение матриц невозможно, то запишите в Z строку:
#   "Упс! Что-то пошло не так..."

# Примечание. В этой задаче не нужно ничего выводить на печать. Только создать Z.

# Про try-except: https://pythonworld.ru/tipy-dannyx-v-python/isklyucheniya-v-python-konstrukciya-try-except-dlya-obrabotki-isklyuchenij.html

import numpy as np

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
B = np.array([
    [11.5, 12.5, 13.5]
])

try:
    Z = np.dot(A,B)
except:
    Z = 'Упс! Что-то пошло не так...'

print(Z)
```
