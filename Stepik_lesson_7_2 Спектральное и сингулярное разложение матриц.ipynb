```python
# Шаг 7_2_8

# Сингулярное разложение матрицы

import numpy as np

np.set_printoptions(precision = 2, suppress = True)

X = np.array([[1, 2, 0], [0, 0, 5], [3, -4, 2], [1, 6, 5], [0, 1, 0]])
X
```

```python
U, s, W = np.linalg.svd(X)
```

```python
U
```

```python
V = W.T
V
```

```python
s
```

```python
D = np.zeros_like(X, dtype = float)
D
```

```python
D[np.diag_indices(min(X.shape))] = s
D
```

```python
np.dot(U.T, U)
```

```python
np.dot(V.T, V)
```

```python
np.dot(U, D)
```

```python
np.dot(np.dot(U, D), V.T)
```

```python
# Шаг 7_2_9

# Связь между спектральным и сингулярным разложением

import numpy as np

np.set_printoptions(precision = 2, suppress = True)

A = np.array([[1.0, 2.0, -1.0], [2.0, 2.0, 5.0], [-1.0, 5.0, -3.0]])

v,w = np.linalg.eigh(A)

D_sp = np.diag(v)
S_sp = w.T

print(S_sp.T, D_sp, S_sp, sep = '\n\n')

```

```python
np.dot(np.dot(S_sp.T, D_sp), S_sp)
```

```python
import numpy as np

np.set_printoptions(precision = 2, suppress = True)

A = np.array([[1.0, 2.0, -1.0], [2.0, 2.0, 5.0], [-1.0, 5.0, -3.0]])

U_sin, s, W = np.linalg.svd(A)
D_sin = np.zeros_like(A, dtype = float)
D_sin[np.diag_indices(min(A.shape))] = s

V_sin = W.T

print(U_sin, D_sin, V_sin.T, sep = '\n\n')
```

```python
np.dot(np.dot(U_sin, D_sin), V_sin.T)
```
