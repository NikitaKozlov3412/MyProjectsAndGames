```python
# Шаг 7_2_2

# Спектральное разложение матрицы

import numpy as np

np.set_printoptions(precision = 2, suppress = True)

A = np.array([[1.0, 2.0, -1.0], [2.0, 2.0, 5.0], [-1.0, 5.0, -3.0]])
A
```

```python
v, w = np.linalg.eigh(A)
```

```python
v
```

```python
w
```

```python
D = np.diag(v)
D
```

```python
S = w.T
S
```

```python
S.T
```

```python
np.dot(S.T, S)
```

```python
np.dot(np.dot(S.T, D), S)
```
