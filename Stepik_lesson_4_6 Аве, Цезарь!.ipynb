```python
# Шаг 4_6_1

# Тема: Шифр Цезаря

# Шифрование — обратимое преобразование информации в целях сокрытия от неавторизованных лиц, с предоставлением доступа авторизованным пользователям.

# Шифровать можно с помощью замены: например, заменяя слова или отдельные символы.
# Можно даже всё сообщение написать непонятными закорючками — или просто "снаряды" заменять на "семечки".

# Другой техникой шифрования с заменой (простой, но гораздо более технологичной) является **шифр Цезаря**.

# В шифре Цезаря:
#   • Каждый символ заменяется на другой по определённому правилу — «сдвигая» его в алфавите.
#   • Сдвиг происходит для каждого символа одинаково.
#   • Если доходим до конца алфавита — начинаем заново.

# Пример: шифр Цезаря со сдвигом 3:
#   A → D
#   B → E
#   C → F
#   ...
#   Z → C (после Z идёт снова A)

# Расшифрование — тоже простое: заменяем символы со сдвигом, но в обратном направлении.

# Преимущество: не надо хранить целый шифроблокнот — достаточно запомнить число (ключ).
# Для алфавитов с чётным числом символов (например, английского) особняком стоит ключ с длиной в половину количества символов алфавита (13 для английского) — такой ключ позволяет зашифровывать и расшифровывать, не меняя знака.

```

```python
# Шаг 4_6_2

# Задание: caesar(text, key)

# Напишите функцию `caesar(text, key)`, возвращающую зашифрованный текст, работающую только с латинским алфавитом.

# Функция принимает:
#   • `text` — исходный текст (который надо зашифровать или расшифровать).
#   • `key` — ключ (сдвиг). Может быть отрицательным или больше 26.

# Алгоритм:
#   • Удаляются все пробелы и знаки препинания.
#   • Все символы переводятся в верхний регистр.
#   • Каждая буква заменяется на другую по правилу "сдвига" в алфавите (A=0, B=1, ..., Z=25).
#   • Если сдвиг выходит за границы алфавита — используется `% 26`.

# Примеры:
#   caesar("Ave, Caesar", 3) → "DYHFDHVDU"
#   caesar("dyhfdhvdu", -3) → "AVECAESAR"

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

def caesar(text, key):
    key = int(key)
    M1 = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    a = list(text.upper())
    while key < 0:
        key += 26
    while key >= 25:
        key -= 26
    c = []
    for i in a:
        if i == ',' or i == '.' or i == '?' or i == '!' or i == ' ' or i == ':' or i == ';' or i == '\'' or i == '"':
            continue
        if i in M1 and (M1.index(i) + key) < 26:
            k = M1.index(i) + key
        else:
            k = M1.index(i) + key - 26
        p = M1[k]
        c.append(p)
    d = ''.join(c)
    return d

caesar("In cryptography, a Caesar cipher, also known as Caesar's cipher, the shift cipher, Caesar's code or Caesar shift, is one of the simplest and most widely known encryption techniques. It is a type of substitution cipher in which each letter in the plaintext is replaced by a letter some fixed number of positions down the alphabet.", 42)
```

```python
# Шаг 4_6_3

# Задание: caesar(text, key, alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ')

# Напишите функцию `caesar(text, key, alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ')`, расширяющую возможности предыдущего шага.

# Функция должна:
#   • Работать с любым переданным алфавитом.
#   • Если алфавит не передан — использовать английский ('ABCDEFGHIJKLMNOPQRSTUVWXYZ').
#   • Удалять из текста все символы, которых нет в алфавите (включая пробелы и знаки препинания).
#   • Переводить весь текст в верхний регистр (если алфавит в верхнем регистре).

# Примеры:
#   caesar("Ave, Caesar", 3) → "DYHFDHVDU"
#   caesar("dyhfdhvdu", -3) → "AVECAESAR"
#   caesar("АБВГД", 1, "АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ") → "БВГДЕ"

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

def caesar(text, key, alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'):
    key = int(key)
    M1 = str(alphabet)
    len_alphabet = len(alphabet)
    a = list(text.upper())
    while key < 0:
        key += len_alphabet
    while key >= (len_alphabet - 1):
        key -= len_alphabet
    c = []
    for i in a:
        if i == ',' or i == '.' or i == '?' or i == '!' or i == ' ' or i == ':' or i == ';' or i == '\'' or i == '"':
            continue
        if i in M1 and (M1.index(i) + key) < len_alphabet:
            k = M1.index(i) + key
        else:
            k = M1.index(i) + key - len_alphabet
        p = M1[k]
        c.append(p)
    d = ''.join(c)
    return d

caesar("А что, если это будет смесь, Runglish? Let me speak from my heart!.. Заметьте, что буквы A и А визуально идентичны, так что на бумаге мы вряд ли расшифровали бы такое...", 13, "ABCDEFGHIJKLMNOPQRSTUVWXYZАБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ")
```
