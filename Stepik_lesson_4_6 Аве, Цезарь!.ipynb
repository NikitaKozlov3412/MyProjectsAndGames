```python
# Шаг 4_6_1

# Тема: Шифр Цезаря

# Шифрование — обратимое преобразование информации в целях сокрытия от неавторизованных лиц, с предоставлением доступа авторизованным пользователям.

# Шифровать можно с помощью замены: например, заменяя слова или отдельные символы.
# Можно даже всё сообщение написать непонятными закорючками — или просто "снаряды" заменять на "семечки".

# Другой техникой шифрования с заменой (простой, но гораздо более технологичной) является **шифр Цезаря**.

# В шифре Цезаря:
#   • Каждый символ заменяется на другой по определённому правилу — «сдвигая» его в алфавите.
#   • Сдвиг происходит для каждого символа одинаково.
#   • Если доходим до конца алфавита — начинаем заново.

# Пример: шифр Цезаря со сдвигом 3:
#   A → D
#   B → E
#   C → F
#   ...
#   Z → C (после Z идёт снова A)

# Расшифрование — тоже простое: заменяем символы со сдвигом, но в обратном направлении.

# Преимущество: не надо хранить целый шифроблокнот — достаточно запомнить число (ключ).
# Для алфавитов с чётным числом символов (например, английского) особняком стоит ключ с длиной в половину количества символов алфавита (13 для английского) — такой ключ позволяет зашифровывать и расшифровывать, не меняя знака.

```

```python
# Шаг 4_6_2

# Задание: caesar(text, key)

# Напишите функцию `caesar(text, key)`, возвращающую зашифрованный текст, работающую только с латинским алфавитом.

# Функция принимает:
#   • `text` — исходный текст (который надо зашифровать или расшифровать).
#   • `key` — ключ (сдвиг). Может быть отрицательным или больше 26.

# Алгоритм:
#   • Удаляются все пробелы и знаки препинания.
#   • Все символы переводятся в верхний регистр.
#   • Каждая буква заменяется на другую по правилу "сдвига" в алфавите (A=0, B=1, ..., Z=25).
#   • Если сдвиг выходит за границы алфавита — используется `% 26`.

# Примеры:
#   caesar("Ave, Caesar", 3) → "DYHFDHVDU"
#   caesar("dyhfdhvdu", -3) → "AVECAESAR"

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

def caesar(text, key):
    key = int(key)
    M1 = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    a = list(text.upper())
    while key < 0:
        key += 26
    while key >= 25:
        key -= 26
    c = []
    for i in a:
        if i == ',' or i == '.' or i == '?' or i == '!' or i == ' ' or i == ':' or i == ';' or i == '\'' or i == '"':
            continue
        if i in M1 and (M1.index(i) + key) < 26:
            k = M1.index(i) + key
        else:
            k = M1.index(i) + key - 26
        p = M1[k]
        c.append(p)
    d = ''.join(c)
    return d

caesar("In cryptography, a Caesar cipher, also known as Caesar's cipher, the shift cipher, Caesar's code or Caesar shift, is one of the simplest and most widely known encryption techniques. It is a type of substitution cipher in which each letter in the plaintext is replaced by a letter some fixed number of positions down the alphabet.", 42)
```

```python
# Шаг 4_6_10

# Задание: Цилиндр Джеферсона

# Напишите функцию `jefferson_encryption(text, discs, step, reverse=False)`, реализующую "цилиндр Джеферсона".

# Функция принимает:
#   • `text` — исходный текст.
#   • `discs` — список из n строк (дисков), где каждый диск — это алфавит в случайном порядке.
#   • `step` — сдвиг, на который смещаются символы с помощью алгоритма Цезаря (ключ шифрования для шифра Цезаря).
#   • `reverse` — признак расшифровки (по умолчанию False — шифровка).

# Алгоритм:
#   1. Создайте n новых алфавитов (дисков) и поместите их в переменную `discs`.
#   2. Для каждого символа текста:
#        - Берём i-й диск (i = позиция символа в тексте % n).
#        - Находим индекс символа в этом диске.
#        - Смещаем индекс на `step` (если `reverse=True` — на `-step`).
#        - Берём новый символ из диска по новому индексу (с учётом цикличности `% len(disc)`).
#   3. Возвращаем зашифрованный текст.

# Примеры:
#   jefferson_encryption("Some encrypted text", discs, 4) → "NUXHUEVGQBIJJZNV"
#   jefferson_encryption("NUXHUEVGQBIJJZNV", discs, 4, reverse=True) → "SOMEENCRYPTEDTEXT"

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.


import random as rnd

rnd.seed(a = 42)

def disc_generator(clear_alphabet, n):
    discs = []
    # c = list(clear_alphabet)
    for i in range(n):
        c = list(clear_alphabet)
        rnd.shuffle(c)
        d = ''.join(c)
        discs.append(d)
        # print(d)
    return discs

disc_generator('ABCDEFGHIJKLMNOPQRSTUVWXYZ', 6)
```

```python
import random as rnd

rnd.seed(42)

def disc_generator(clear_alphabet, n):
    discs = []
    for _ in range(n):
        c = list(clear_alphabet)
        rnd.shuffle(c)
        discs.append(''.join(c))
    return discs

def jefferson_encryption(text, discs, step, reverse=False):
    # Очищаем текст: оставляем только символы из первого диска
    clean_text = ''.join(char.upper() for char in text if char.upper() in discs[0])

    result = []
    n_discs = len(discs)  # количество дисков

    for i, char in enumerate(clean_text):
        disk = discs[i % n_discs]      # выбираем диск циклически
        pos = disk.index(char)        # позиция символа на диске
        shift = -step if reverse else step  # направление сдвига
        new_pos = (pos + shift) % len(disk)  # новая позиция (с учётом цикла)
        result.append(disk[new_pos])   # новый символ

    return ''.join(result)

dg = disc_generator('АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ', 36)
jefferson_encryption('Съешь еще этих мягких французских булок. Кстати, в этом тексте пришлось заменить одну букву...', dg, 4, True)
```
