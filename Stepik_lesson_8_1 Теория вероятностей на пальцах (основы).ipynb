```python
# Шаг 8_1_7

# Задача 3.1

# Лента длиной L сантиметров разрезается на 2 куска в случайном месте.

# Найдите вероятность того, что один из кусков имеет длину не менее M сантиметров.

# На вход подаётся 2 числа: L, M

# Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.

# Примечание. Задача легко решается аналитически, поэтому за неё не даются баллы, но всё же попробуйте решить её с помощью метода Монте-Карло.

# Sample Input 1:
# 100
# 80

# Sample Output 1:
# 39.9%

# Sample Input 2:
# 200
# 100

# Sample Output 2:
# 100.0%

# Sample Input 3:
# 100
# 10

# Sample Output 3:
# 100.0%

L = int(input())
M = int(input())


import random

num = 1000000
suc = 0

for _ in range(num):
        edge_1 = random.uniform(0, L)
        edge_2 = L - edge_1
        if edge_1 >= M or edge_2 >= M:
            suc += 1


# print(edge_1)
# print(edge_2)


chance = suc/num

print(round(chance*100, 1), '%', sep = '')

```

```python
# Шаг 8_1_8

# Задача 3.2

# Лента длиной L сантиметров разрезается на N кусков в случайных местах (т.е. делают N-1 разрез).

# Найдите вероятность того, что один из кусков имеет длину не менее M сантиметров.

# На вход подаётся 3 числа: L, N, M

# Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.

# Sample Input 1:
# 100
# 2
# 80

# Sample Output 1:
# 39.9%

# Sample Input 2:
# 2
# 2
# 1

# Sample Output 2:
# 100.0%

# Sample Input 3:
# 300
# 3
# 100

# Sample Output 3:
# 100.0%

# Sample Input 4:
# 100
# 3
# 60

# Sample Output 4:
# 48.0%

```

```python
L = int(input())
N = int(input())
M = int(input())


import random


num = 1000000
suc = 0

for _ in range(num):

    L0 = L
    a = []

    for _ in range(N-1):
        x = random.uniform(0, L0)
        a.append(x)
        # L = L - x
    a.append(L)

    # Сотрировка оказывается влияет на вероятность
    a.sort()

    pieces = [0] + a + [L0]

    tot = []

    for i in range(len(pieces) - 1):
        piece_len = pieces[i+1] - pieces[i]
        tot.append(piece_len)

    found = False
    for j in range(len(tot)):
         if tot[j] >= M:
            found = True
            break

    if found:
        suc += 1


chance = suc/num

print(round(chance*100, 1), '%', sep = '')

```

```python
# Шаг 8_1_9

# Задача 4.1

# В трамвайном парке имеются N трамваев маршрута №1 и M трамваев маршрута №2.
# Какова вероятность того, что вторым по счёту на линию выйдет трамвай маршрута №1?

# На вход подаётся 2 числа: N, M

# Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.

# Sample Input 1:
# 15
# 10

# Sample Output 1:
# 60.0%

# Sample Input 2:
# 10
# 10

# Sample Output 2:
# 50.1%   # (из-за дискретности и метода Монте-Карло — в аналитике будет ровно 50.0%, но в симуляции возможны небольшие флуктуации)

# Sample Input 3:
# 0
# 5

# Sample Output 3:
# 0.0%

# Sample Input 4:
# 1
# 2

# Sample Output 4:
# 33.3%

# Sample Input 5:
# 2
# 0

# Sample Output 5:
# 100.0%

# Аналитический метод

N = int(input())
M = int(input())

P = 100*((N/(N+M))*((N-1)/(N+M-1)) + (M/(N+M))*((N)/(N+M-1)))
print(round(P, 1), '%', sep = '')
```

```python
# Метод Монте-Карло

N = int(input())
M = int(input())

import random

trial = 1000000

tramvai = [1] * N + [2] * M
suc = 0

for _ in range(trial):
    random.shuffle(tramvai)
    if tramvai[1] == 1:
        suc += 1

chance = 100*suc/trial
print(round(chance, 1), '%', sep = '')

```

```python
# Шаг 8_1_10

# Задача 4.2

# В трамвайном парке имеются N трамваев маршрута №1 и M трамваев маршрута №2.
# Какова вероятность того, что K-м по счёту на линию выйдет трамвай маршрута №1?

# На вход подаётся 3 числа: N, M, K

# Выведите на печать вероятность в процентах, округлив до 1 знака после запятой.

# Sample Input 1:
# 15
# 10
# 2

# Sample Output 1:
# 59.9%

# Sample Input 2:
# 10
# 10
# 2

# Sample Output 2:
# 50.1%

# Sample Input 3:
# 15
# 10
# 3

# Sample Output 3:
# 60.1%

# Sample Input 4:
# 4
# 1
# 5

# Sample Output 4:
# 80.0%

# Sample Input 5:
# 4
# 0
# 4

# Sample Output 5:
# 100.0%


```

```python
N = int(input())
M = int(input())
K = int(input())

import random

trial = 1000000

tramvai = [1] * N + [2] * M
suc = 0

for _ in range(trial):
    random.shuffle(tramvai)
    if tramvai[K-1] == 1:
        suc += 1

chance = 100*suc/trial
print(round(chance, 1), '%', sep = '')
```
