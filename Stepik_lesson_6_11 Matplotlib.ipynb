```python
# Шаг 6_11_2

import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0, 5, 0.1)
y = x**2
```

```python
plt.figure()
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
```

```python
plt.figure()
plt.plot(x,y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
plt.show()
```

```python
# Python 2 без буквы u в начале ругается, если я верно понял, но в Python 3 работает нормально:
plt.figure()
plt.plot(x,y, 'r')
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.title('Заголовок')
plt.show()
```

```python
plt.figure()
plt.plot(x,y, 'r')
plt.xlabel(u'Абсцисса')
plt.ylabel(u'Ордината')
plt.title(u'Заголовок')
plt.show()
```

```python
x = range(-10, 10)
```

```python
y = [element**2 for element in x]
```

```python
print(x)
print(y)
```

```python
plt.figure()
plt.plot(x,y, 'r')
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.title('Заголовок')
plt.show()
```

```python
plt.figure()
plt.plot(x,y, 'r', y,x, 'g')
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.title('Заголовок')
plt.show()
```

```python
plt.figure()
plt.plot(x,y, 'r')
plt.plot(y,x)
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.title('Заголовок')
plt.show()
```

```python
plt.figure(figsize = (13,8)) # Размеры в дюймах
plt.plot(x,y, 'r')
plt.plot(y,x)
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.title('Заголовок')
plt.show()
```

```python
plt.figure(figsize = (13,8), dpi = 120) # Размеры в дюймах # dpi - разрешение
plt.plot(x,y, 'r')
plt.plot(y,x)
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.title('Заголовок')
plt.show()
```

```python
plt.figure(figsize = (13,8), dpi = 120) # Размеры в дюймах # dpi - разрешение
plt.plot(x,y, 'r')
plt.plot(y,x)
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')
plt.title('Заголовок')
# вместо show сохраняем на диск, график всё равно отобразится:
plt.savefig('1.png')
```
