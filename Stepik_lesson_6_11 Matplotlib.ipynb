```python
# Шаг 6_11_5

# Легенда, размер шрифта

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.arange(0.1, 5, 0.1)

def f(x):
    return x**2

def g(x):
    return 1 / x

def h(x):
    return x*np.sin(x)

```

```python
fig, axes = plt.subplots(figsize = (13,8))
axes.plot(x, f(x), x, g(x))
axes.plot(x, h(x))
axes.legend(['f(x)', 'g(x)', 'h(x)'])
```

```python
fig, axes = plt.subplots(figsize = (13,8))
axes.plot(x, f(x), label = u'функция f(x)')
axes.plot(x, g(x))
axes.plot(x, h(x), label = u'функция h(x)')
axes.legend()
```

```python
fig, axes = plt.subplots(figsize = (13,8))
axes.plot(x, f(x), label = u'функция f(x)')
axes.plot(x, g(x))
axes.plot(x, h(x), label = u'функция h(x)')
axes.legend(loc = 5) # местоположение легенды задаётся параметром loc
```

```python
fig, axes = plt.subplots(figsize = (13,8))
axes.plot(x, f(x), label = '$x^2$') # В названия легенд можно писать формулы, заключённые в символы "доллары"
axes.plot(x, g(x), label = '$\\frac{1}{x}$')
axes.plot(x, h(x), label = r'$x \cdot sin(x)$')
axes.legend(loc = 5) # местоположение легенды задаётся параметром loc
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# Диапазоны x без нуля
x_neg = np.linspace(-5, -0.1, 500)
x_pos = np.linspace(0.1, 5, 500)

def f(x, alpha, beta):
    return 1 + (alpha / x) ** beta

params = [
    (1, 1),   # красный
    (2, 1),   # синий
    (1, 2)    # зелёный
]

# Явно задаём цвета — по одному на каждую пару
colors = ['red', 'blue', 'green']

plt.figure(figsize=(12, 8))

for i, (alpha, beta) in enumerate(params):
    color = colors[i]
    y_neg = f(x_neg, alpha, beta)
    y_pos = f(x_pos, alpha, beta)

    # Рисуем отрицательную часть
    plt.plot(x_neg, y_neg, color=color, linewidth=2, label=rf'$\alpha={alpha},\ \beta={beta}$')
    # Рисуем положительную часть — тем же цветом!
    plt.plot(x_pos, y_pos, color=color, linewidth=2)

plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.title(r'График функции $f(x) = 1 + \left(\frac{\alpha}{x}\right)^\beta$ при различных $\alpha$ и $\beta$', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.ylim(-20, 30)
plt.axhline(0, color='brown', linewidth=0.5, linestyle='--')
plt.axvline(0, color='brown', linewidth=0.5, linestyle='--')
plt.savefig('1_(ax)b.svg')
```
