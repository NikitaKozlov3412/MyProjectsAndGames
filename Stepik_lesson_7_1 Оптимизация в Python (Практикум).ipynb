```python
# Шаг 7_1_5

# N друзей шестиугольника - усечённый икосаэдр

# Шестиугольник из задачи "9 друзей шестиугольника" завязал с тёмными делишками, занялся честным бизнесом и обзавёлся новыми друзьями.
# Теперь они не только классно смотрятся на плоскости, но и могут быть "свёрнуты" в объёмную фигуру — усечённый икосаэдр (математическую абстракцию для футбольного мяча).

# Плоская фигура состоит из:
# - 20 шестиугольников
# - 12 пятиугольников
# с одинаковой длиной ребра a.

# Напишите 2 функции:
# - S(x), принимающую на вход натуральное число (длину ребра шестиугольника), а возвращающую площадь плоской фигуры
# - S_ceil(x), принимающую на вход натуральное число (длину ребра шестиугольника), а возвращающую площадь плоской фигуры, округлённую "вверх"

# Примечание 1. В этой задаче не требуется считывать данные или выводить их на печать. Необходимо только реализовать функции S(x) и S_ceil(x).
# Входные и выходные данные даны только для самоконтроля!

# Примечание 2.
# - Площадь правильного пятиугольника
# - Площадь правильного шестиугольника

# Примечание 3. Объёмная фигура из этой задачи имеет ряд интересных свойств, которые изучаются в курсе "Математика для всех" на 2 неделе.

# Sample Input 1:
# 7

# Sample Output 1:
# 3557.7553986725625 3558

# Sample Input 2:
# 5

# Sample Output 2:
# 1815.181325853348 1816

```

```python
import math as mt

x = int(input())

def S(x):
    S5 = (mt.sqrt(5) * (mt.sqrt(5 + 2 * mt.sqrt(5))))/4 * x**2
    S6 = (3 * mt.sqrt(3)) /2 * x**2
    Stot = 20*S6 + 12*S5
    return Stot

def S_ceil(x):
    S5 = (mt.sqrt(5) * (mt.sqrt(5 + 2 * mt.sqrt(5))))/4 * x**2
    S6 = (3 * mt.sqrt(3)) /2 * x**2
    Stot = 20*S6 + 12*S5
    return mt.ceil(Stot)

print(S(x), S_ceil(x))
```

```python
# Шаг 7_1_9

# Поиск минимума функции с помощью scipy.optimize.golden()

# scipy.optimize.golden() - находит минимум функции. Как было показано в конце предыдущего шага, если не передавать в неё никаких параметров кроме функции, минимум которой ищется, то возвращён будет аргумент, при котором этот минимум достигается. Более подробно с функцией можно ознакомиться в документации.

# 1) Реализуйте 2 функции:
#   - f(x) = (x + a)² - b
#   - g(x) = |f(x)|

# Примечание. Считайте, что переменные a и b заданы выше в коде, не надо передавать их параметры функции.

# 2) Импортируйте функцию golden из модуля scipy.optimize.

# 3) Найдите аргументы, при которых достигается минимум функций f(x) и g(x), ответ запишите через пробел.

# Примечание. Пример дан для a = 3, b = 4

# Sample Input:

# Sample Output:
# -3.0 -0.99999999987819608

from scipy.optimize import golden

def f(x):
    return (x+a)**2 - b

def g(x):
    return abs(f(x))

min_f = golden(f)
min_g = golden(g)

print(min_f, min_g)
```
