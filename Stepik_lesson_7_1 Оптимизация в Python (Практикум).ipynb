```python
# Шаг 7_1_2

# Несколько важных замечаний

# О функциях

# Про использование функций внутри функций

# В Python нет ограничений на использование функций внутри других функций.

# В данном случае:

# def f(x):
#     return x**3
#
# def g(x):
#     return abs(f(x))
#
# def k(x):
#     return 2*x + f(x) + g(x)

# print('f(-1):', f(-1))
# print('g(-1):', g(-1))
# print('k(-1):', k(-1))
#
# f(-1): -1
# g(-1): 1
# k(-1): -2

# print('f(-3):', f(-3))
# print('g(-3):', g(-3))
# print('k(-3):', k(-3))
#
# f(-3): -27
# g(-3): 27
# k(-3): -6

# k(x) = 2x + f(x) + g(x) = 2x + f(x) + |f(x)| = 2x + x^3 + |x^3|

# Легко заметить, что значения f(x) и g(x) будут противоположны по знаку, если x < 0 и одинаковые, если x > 0.

# Так же легко заметить, что k(x) сводится просто к функции 2x, если x < 0, и более сложной 2x + 2x^3, если x > 0.

# На примере выше мы в этом можем убедиться экспериментально.
```

```python
# Шаг 7_1_3

# Напишите функцию f(x) = x² + 2x + 1

# Функция должна принимать на вход действительное число и возвращать действительное число.

# Примечание. В данной задаче не надо ничего выводить на печать! Ваша задача реализовать только функцию f(x).

x = float(input())

def g(x):
    return x**2

def f(x):
    return g(x) + 2*x + 1

print(f(x))
```

```python
# Шаг 7_1_4

# Напишите функцию g(x), которая будет возвращать только целую часть от f(x).

# f(x) уже определена выше по коду и не известна вам, однако гарантируется, что она возвращает число типа float.

# Примечание. В этой задаче не требуется считывать данные или выводить их на печать. Необходимо только реализовать функцию g(x).
# Для тренировки можете использовать функцию из предыдущего задания в качестве f(x).

x = float(input())

def f(x):
    return x**3

def g(x):
    return int(f(x))

print(g(x))
```

```python
x = float(input())

def f(x):
    return x**3

def g(x):
    return f(x)//1

print(g(x))
```

```python
import math as mt

x = float(input())

def f(x):
    return x**3

def g(x):
    return mt.trunc(f(x))

print(g(x))
```
