```python
# Шаг 4_2_7

# Задание: kaprekar_check(n)

# Напишите функцию `kaprekar_check(n)`, принимающую на вход одно натуральное число,
# а возвращающую логическое значение (`True` или `False`) — возможно ли для этого числа запустить процесс Капреакара.

# Критерии возможности:
#   • Число должно быть 3-, 4- или 6-значным.
#   • Не все цифры в числе должны быть одинаковыми.
#   • Число не должно равняться 100, 1000 или 100000.

# Рекомендация:
#   Используйте функцию `numerics(n)` совместно с множествами (`set()`) для определения количества уникальных цифр.

# Примеры:
#   kaprekar_check(4086) → True
#   kaprekar_check(486) → True
#   kaprekar_check(408886) → True
#   kaprekar_check(86) → False (не 3, 4 или 6 знаков)
#   kaprekar_check(1234567) → False (7 знаков)
#   kaprekar_check(100) → False (запрещено)
#   kaprekar_check(1000) → False (запрещено)
#   kaprekar_check(1111) → False (все цифры одинаковые)

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

n = int(input())

def kaprekar_check(n):
    b = str(n)
    c = []
    for i in b:
        c.append(int(i))
    d = set(c)
    if len(b) != 3 and len(b) != 4 and len(b) != 6:
        return False
    elif len(d) < 2:
        return False
    elif n == 100 or n == 1000 or n == 100000:
        return False
    else:
        return True

kaprekar_check(n)
```

```python
# Шаг 4_2_8

# Задание: kaprekar_step(L) ver.2

# Обновите функцию `kaprekar_step(L)` — теперь она должна работать с числами **любой длины** (не только 4-значными).

# Напишите функцию `kaprekar_step(L)`, принимающую на вход список натуральных чисел (цифр, из которых состоит число),
# проводит 1 шаг сходимости к постоянной Капреакара и возвращает получившееся число.

# Функция должна:
#   1. Сформировать внутри два числа:
#        • наибольшее — цифры в порядке убывания,
#        • наименьшее — цифры в порядке возрастания.
#   2. Вычислить разницу: наибольшее - наименьшее.
#   3. Вернуть результат.

# Примеры:
#   kaprekar_step([5, 6, 3]) → 297
#   kaprekar_step([2, 7, 1, 8]) → 7443
#   kaprekar_step([5, 4, 9, 9, 4, 5]) → 549945

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

L = [5, 4, 9, 9, 4, 5]

def kaprekar_step(L):
    l = []
    for i in L:
        l.append(str(i))
    str_asc = sorted(l)
    # str_desc = reversed(str_asc)
    str_desc = str_asc[::-1]
    # str_desc = sorted(l, reverse = True)
    str_min = ''.join(str_asc)
    str_max = ''.join(str_desc)
    alpha = int(str_min)
    gamma = int(str_max)
    delta = gamma - alpha
    return delta

kaprekar_step(L)
```
