```python
# Шаг 4_2_3

# Задание: kaprekar_step(L)

# Напишите функцию `kaprekar_step(L)`, принимающую на вход список натуральных чисел (цифр, из которых состоит число, проходящее через процесс Капреакара),
# проводит 1 шаг сходимости к постоянной Капреакара и возвращает 4-значное число.

# Функция должна:
#   1. Сформировать внутри два числа:
#        • наибольшее — цифры в порядке убывания,
#        • наименьшее — цифры в порядке возрастания.
#   2. Вычислить разницу: наибольшее - наименьшее.
#   3. Вернуть результат.

# Примеры:
#   kaprekar_step([4, 5, 6, 8]) → 4086
#   kaprekar_step([6, 8, 0, 4]) → 8172
#   kaprekar_step([2, 7, 1, 8]) → 7443

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.


L = [2, 7, 1, 8]

def kaprekar_step(L):
    l = [str(t) for t in L]
    min = int(''.join(sorted(l)))
    max = int(''.join(sorted(l, reverse= True)))
    p = max - min
    return p


kaprekar_step(L)
```
