```python
# Шаг 4_7_2

# Задание: rotor(symbol, n, reverse=False)

# Напишите функцию `rotor(symbol, n, reverse=False)`, реализующую шифрование одним ротором Энигмы.

# Функция принимает:
#   • `symbol` — символ, поступающий для шифрования.
#   • `n` — номер ротора (0, 1, 2, 3).
#   • `reverse` — признак обратного направления (по умолчанию False — прямое).

# Возвращает строку с зашифрованным символом.

# Алгоритм:
#   • Если `n = 0` — "не шифрованный ротор" — возвращает тот же символ.
#   • Для `n = 1, 2, 3` — использует таблицу соответствия из спецификации:
#         Rotor I:  E K M F L G D Q V Z N T O W Y H X U S P A I B R C J
#         Rotor II: A J D K S I R U X B L H W T M C Q G Z N P Y F V O E
#         Rotor III: B D F H J L C P R T X V Z N Y E I W G A K M U S Q O

#   • При `reverse=True` — используется обратное соответствие (например, если 'A' → 'E', то 'E' → 'A').

# Примеры:
#   rotor('A', 0) → 'A'
#   rotor('A', 1) → 'E'
#   rotor('E', 1, reverse=True) → 'A'

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

RDICT = {1: ('AELTPHQXRU', 'BKNW', 'CMOY', 'DFG', 'IV', 'JZ', 'S'),
         2: ('FIXVYOMW', 'CDKLHUP', 'ESZ', 'BJ', 'GR', 'NT', 'A', 'Q'),
         3: ('ABDHPEJT', 'CFLVMZOYQIRWUKXSG', 'N'),
         4: ('AEPLIYWCOXMRFZBSTGJQNH', 'DV', 'KU'),
         5: ('AVOLDRWFIUQ', 'BZKSMNHYC', 'EGTJPX'),
         6: ('AJQDVLEOZWIYTS', 'CGMNHFUX', 'BPRK'),
         7: ('ANOUPFRIMBZTLWKSVEGCJYDHXQ'),
         8: ('AFLSETWUNDHOZVICQ', 'BKJ', 'GXY', 'MPR'),
         'beta': ('ALBEVFCYODJWUGNMQTZSKPR', 'HIX'),
         'gamma': ('AFNIRLBSQWVXGUZDKMTPCOYJHE'),
         }

RDICT_0 = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

def rotor (symbol, n, reverse = False):
    text = []
    for aa in symbol:
        if n == 0:
            if aa.upper() in RDICT_0:
                text.append(aa.upper())
        else:
            tuple_dict = list(RDICT[n])
            # print(tuple_dict)
            aa = aa.upper()
            k = len(tuple_dict)
            # print(k)
            for i in range(k):
                if aa in tuple_dict[i]:
                    # print(tuple_dict[i])
                    idx = tuple_dict[i].index(aa)
                    step = 1
                    if reverse:
                        step = -1
                    if idx + step >= len(tuple_dict[i]):
                        idx_n = idx + step - len(tuple_dict[i])
                    elif idx + step < 0:
                        idx_n = idx + step + len(tuple_dict[i])
                    else:
                        idx_n = idx + step
                    bb = tuple_dict[i][idx_n]
                    # print(bb)
                    text.append(bb)
    total_text = ''.join(text)
    return total_text


rotor('SOME ENCRYPTED TEXT FOR EXAMPLE', 3, True)
```

```python
# Шаг 4_7_3

# Задание: reflector(symbol, n)

# Напишите функцию `reflector(symbol, n)`, реализующую отражатель Энигмы.

# Функция принимает:
#   • `symbol` — символ, поступающий для шифрования.
#   • `n` — номер отражателя (0 или 1).

# Возвращает строку с зашифрованным символом.

# Алгоритм:
#   • Если `n = 0` — "без шифрования" — возвращает тот же символ.
#   • Если `n = 1` — использует таблицу соответствия из спецификации:
#         Reflector B: Y R U H Q S L D P X N G O K M I E B F Z C W V J A T

# Примеры:
#   reflector('A', 0) → 'A'
#   reflector('A', 1) → 'T'
#   reflector('T', 1) → 'A' (обратимость)

# Примечание:
#   • В этой задаче не нужно ничего считывать и ничего выводить на печать.
#   • Требуется только реализовать функцию.

REFLECTORS = {0: 'ABCDEFGHIJKLMNOPQRSTUVWXYZ',
              1: 'YRUHQSLDPXNGOKMIEBFZCWVJAT',
              2: 'FVPJIAOYEDRZXWGCTKUQSBNMHL',
              3: 'ENKQAUYWJICOPBLMDXZVFTHRGS',
              4: 'RDOBJNTKVEHMLFCWZAXGYIPSUQ',
              }

def reflector(symbol, n):
    text = []
    list_dict = list(REFLECTORS[n])
    list_0 = list(REFLECTORS[0])
    # print(list_dict)
    for aa in symbol:
        aa = aa.upper()
        if aa in list_dict:
            k = list_0.index(aa)
            d = list_dict[k]
            text.append(d)
            # print(aa, k, d)
    total_text = ''.join(text)
    return total_text

reflector(' SOME ENCRYPTED TEXT FOR EXAMPLE', 1)
```
